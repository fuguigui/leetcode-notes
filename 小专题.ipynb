# 子序列问题

[labuladong](https://labuladong.github.io/algo/3/24/78/)

一旦涉及到子序列和最值，那几乎可以肯定，考察的是动态规划技巧，时间复杂度一般都是 O(n^2)。

- [(medium)300 Longest Increasing Subsequence](https://leetcode.cn/problems/longest-increasing-subsequence/)
    最长递增子序列。限制是相对顺序。
    - 应用: [(hard)354 Russian Doll Envelopes](https://leetcode.cn/problems/russian-doll-envelopes/)。多了一步是，对w按升序排序，按h降序排序。

```python
# 300
import bisect
class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        len2min = [nums[0]]
        for num in nums[1:]:
            if num > len2min[-1]:
                len2min.append(num)
            elif num == len2min[-1]:
                continue
            else:
                num_id = bisect.bisect_left(len2min, num)
                len2min[num_id] = num
        return len(len2min)
```

```python
# 354
import bisect
class Solution:
    def maxEnvelopes(self, envelopes: List[List[int]]) -> int:
        envelopes.sort(key=lambda x: (x[0], -x[1]))
        len2min = [envelopes[0][1]]
        for env in envelopes:
            v = env[1]
            if v > len2min[-1]:
                len2min.append(v)
            elif v < len2min[-1]:
                insert_id = bisect.bisect_left(len2min, v)
                len2min[insert_id] = v
        return len(len2min)
```

# 滑动窗口

常见问题：

```python
# 239 滑动窗口模板+新滑动窗口设计
# 执行用时：556 ms, 在所有 Python3 提交中击败了50.81% 的用户
# 内存消耗：28.3 MB, 在所有 Python3 提交中击败了35.06% 的用户
from collections import deque
class MaxWindow:
    def __init__(self, k):
        self.window = deque()
        self.k = k

    def push(self, n):
        while self.window and self.window[-1] < n:  # 这里用小于，而不是小于等于
            self.window.pop()
        self.window.append(n)

    def top(self):
        return self.window[0]

    def pop(self, left):
        if self.window[0] == left:
            self.window.popleft()

class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        ans = []
        max_window = MaxWindow(k)
        for i, num in enumerate(nums):
            if i < k - 1:
                max_window.push(num)
            else:
                max_window.push(num)
                ans.append(max_window.top())
                max_window.pop(nums[i-k+1])
        return ans
```

```python
# 239 pop不带参数
from collections import deque
class MaxWindow:
    def __init__(self, k):
        self.window = deque()
        self.k = k
        self.cnt = 0 # 自带计数，用作id

    def push(self, n):
        while self.window and self.window[-1][0] < n:  # 这里用小于，而不是小于等于
            self.window.pop()
        self.window.append((n, self.cnt))
        self.cnt += 1

    def top(self):
        return self.window[0][0]

    def pop(self):
        if self.window[0][1] == self.cnt - self.k: # 判断是不是k个之前的id
            self.window.popleft()

class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        ans = []
        max_window = MaxWindow(k)
        for i, num in enumerate(nums):
            if i < k - 1:
                max_window.push(num)
            else:
                max_window.push(num)
                ans.append(max_window.top())
                max_window.pop()
                
        return ans
```

- [(medium)2128 Remove All Ones With Row and Column Flips](https://leetcode.cn/problems/remove-all-ones-with-row-and-column-flips)
    - 思路：遍历，一共m+n种操作，需要的话都做一遍，最后看看成不成。有一个小trick是，对于每一行第一个数，如果是1，那么要么需要被该行翻转，要么需要被第一列翻转。
        - 对于左上角的数，
            - 如果是1，在翻转第一行和第一列之间，只能选择一个。默认选择第一行，因为选择第一列的话，其实是等价的。
            - 如果是0，要么第一行第一列都不翻转，要么都翻转。如果都翻转的话，假设第一行还有其他列有0，翻转之后会变为1，需要翻转对应列；第一列同理。如果都不翻转的话，也同理。总结而言，每行每列开头的元素，决定了要不要翻转该行该列。
