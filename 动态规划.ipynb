# 知识点

动态规划的整体指导思想是**减少无用或重复的计算**。
这种方法的本质是，把一个规模比较大的问题，通过规模比较小的若干问题的结果来得到的

- [X] 什么时候可以使用动态规划来解题？

通常我们要从**有无后效性**入手分析。如果对于某个状态，我们可以只关注状态的值，而不需要关注状态是如何转移过来的话，那么这就是一个无后效性的问题，可以考虑使用 DP 解决。

另外一个更加实在的技巧，我们还可以通过 数据范围 来猜测是不是可以用 DP 来做。

因为 DP 是一个递推的过程，
- 如果数据范围是10^5-10^6的话，可以考虑是不是可以使用一维 DP 来解决
- 如果数据范围是10^2-10^3的话，可以考虑是不是可以使用二维 DP 来解决


将所做过的动态规划问题按照这些类别进行归类，**分析状态的表示方法和状态转移方程的构造方法**在每种类型中的近似之处，会让你更快的学会动态规划。

## 动态规划模板

动态规划和递归（带状态）可以说是一码事，都是在所有的状态空间中寻找一种搜索方法，来避开“暴力穷举”。
- 「自顶向下」进行「递归」求解，
- 动态规划：「自底向上」进行「递推」求解。

动态规划三要素：
- 具备“最优子结构”：可以从子问题的最优结果推出更大规模问题的最优结果。要符合「最优子结构」，子问题间必须互相独立。
- 存在“重叠子问题”：使用「备忘录」或者「DP table」来优化穷举过程，避免不必要的计算。
- 正确的状态转移方程：就是对最后一步的分情况讨论

### 最优子结构


规定的是子问题与原问题的关系。

如果可以把这个问题分解成多个子问题，然后递归地找到每个子问题的最优解，最后通过一定的数学方法对各个子问题的最优解进行组合得出最终的结果。总结来说就是一个问题的最优解是由它的各个子问题的最优解决定的。

- [X] 什么叫做子问题之间互相独立？
想满足最优子结构，子问题之间必须互相独立。如果原问题的解可能出现在两个子问题之间，显然子问题不独立。比如：知道每个班级的最大分数差，求全校的最大分数差。不满足最优子结构，因为全校的最大分数差，可能出现在两个班级之间。

- [X] 遇到这种最优子结构失效情况，怎么办？
策略是：改造问题，**把问题等价转化**。比如全校分数差问题。改造问题，也就是最大分数差，不就等价于最高分数和最低分数的差么，那不就是要求最高和最低分数么。

**将子问题的解进行组合可以得到原问题的解是动态规划可行性的关键**。


最优子结构并不是动态规划独有的一种性质，能求最值的问题大部分都具有这个性质；但反过来，最优子结构性质作为动态规划问题的必要条件，一定是让你求最值的，以后碰到那种恶心人的最值题，思路往动态规划想就对了，这就是套路。


### 重复子问题

重复子问题规定的是子问题与子问题的关系。

重叠的子问题使得直接递归会有很多重复计算，于是就想到**记忆化递归法(其实就是建备忘录)**：若能事先确定子问题的范围就可以建表存储子问题的答案。

动态规划算法中关于最优子结构和重复子问题的理解的关键点

1. 证明问题的方案中包含一种选择，选择之后留下一个或多个子问题
  
2. 设计子问题的递归描述方式
  
3. 证明对原问题的最优解包括了对所有子问题的最优解
  
4. 证明子问题是重叠的（这一步不是动态规划正确性必需的，但是如果子问题无重叠，则效率与一般递归是相同的）



### 状态转移方程

**状态转移方程就是对最后一步的分情况讨论**。

在解题中一般用状态转移方程描述这种组合。

例如原问题的解为 $f(n)$，其中 $f(n)$也叫状态。状态转移方程 $f(n)=f(n−1)+f(n−2)$描述了一种原问题与子问题的组合关系 。

如果猜了一个状态定义，然后发现无法列出涵盖所有情况（不漏）的状态转移方程，多半就是状态定义猜错了，赶紧换个思路，而不是去死磕状态转移方程。

我们的状态转移是要做到「不漏」还是「不重不漏」取决于问题本身：
- 如果是求最值的话，我们只需要确保「不漏」即可，因为重复不影响结果。
- 如果是求方案数的话，我们需要确保「不重不漏」。



从下面的4个要素去逐步剖析解决这道题：

1. 状态是什么：确定dp数组（dp table）以及下标的含义
2. 状态转移方程是什么：确定递推公式
3. 状态的初始值是什么：一些情况是递推公式决定了dp数组要如何初始化！
4. 问题要求的最后答案是什么


每个步骤分析完成之后，就基本上解决了整道动态规划的问题。

### dp数组的遍历方向
拿二维 dp 数组来举例，有时候我们是正向遍历，有时候我们反向遍历，有时候可能会斜向遍历。

只要把住两点就行了：

1. 遍历的过程中，所需的状态必须是已经计算出来的。

2. 遍历结束后，存储结果的那个位置必须已经被计算出来。

## 复杂度分析

对于动态规划的复杂度/计算量分析，有多少个状态，复杂度/计算量就是多少。

因此一维 DP 的复杂度通常是线性的，而二维 DP 的复杂度通常是平方的 。

## 应用技巧

### Debug的方式

debug的最好方式就是把dp数组打印出来，看看究竟是不是按照自己思路推导的！

- 如果打印出来和自己预先模拟推导是一样的，那么就是自己的递归公式、初始化或者遍历顺序有问题了。
- 如果和自己预先模拟推导的不一样，那么就是代码实现细节有问题。

debug时，可以思考这三个问题：

- 这道题目我举例推导状态转移公式了么？
- 我打印dp数组的日志了么？
- 打印出来了dp数组和我想的一样么？


### 优化的方向

- 时间上的优化：
    - 通过变换状态优化：需要一些经验和灵感
    - 通过决策单调优化：比较简单，但适用范围不广，一般只适用于划分型动态规划当中，通常这个方法可以将复杂度降低一个数量级。
    
- 空间上的优化：使用滚动数组进行优化：第i层的状态，已经和第i-2层的状态没有关系了，那么这种情况下，用于存储第i-2层的空间就可以被重复利用

空间压缩技巧：**你看它的状态转移方程，如果计算状态 dp[i][j] 需要的都是 dp[i][j] 相邻的状态，那么就可以使用空间压缩技巧**.来自[链接](https://labuladong.github.io/algo/3/23/70/)


## 分类

### 背包问题

背包问题是「动态规划」中十分经典的一类问题，背包问题本质上属于组合优化的**NP完全问题**。

- 可以将「NP完全问题」简单理解为「无法直接求解」的问题。只能通过「穷举」+「验证」的方式进行求解。

「背包问题」大概是对应这样的一类问题：

泛指一类**给定价值与成本**，同时**限定决策规则**，在这样的条件下，如何实现**价值最大化**的问题。

例如：「背包」是指给定物品**价值与体积**（对应了「给定价值与成本」），在**规定容量**下（对应了「限定决策规则」）如何使得**所选物品的总价值最大**。

- 01背包：每个物品最多选择一次.状态转移方程是
$
dp(i,j) = max(dp(i-1,j), dp(i-1, j-vi)+wi)
$

- 完全背包：每个物品可以选择无穷多次。有n种物品和一个容量为C的背包，每种物品都有无限件。第i件物品的体积是vi，价值是wi。求解将哪些物品装入背包可使这些物品的费用总和不超过背包容量，且价值总和最大。状态转移方程是
$
dp(i,j) = max(dp(i-1,j), dp(i, j-vi)+wi)
$

区别在于第二项，完全背包依然是看i，而01背包看的是i-1。两者都可以压缩为1维空间，
- 01背包依赖的是上一层i-1，前面的j-vi的值，所以要从大到小遍历，以确保i-1的j-vi没有被更新成第i行的。
- 而完全背包依赖的是本层i，前面的j-vi的值，所以要从小到大遍历，以确保i的j-vi已经被更新成第i行的。


- 多重背包：每个物品的数量有限。优化方式：
    - 扁平化展开指的是对物品数量si进行拆分，拆分成多个数量为1的物品，转换成了「01 背包」。
        - 最直接的拆分方式：就是拆分成si个提及为vi，价值为wi的物品。这样拆分的不好在于，如果原来取了一个该物品，在新的拆分下会对应到si种情况，即分别取第1个、第2个。。。第si个。增加了枚举的复杂度。整体复杂度是$O(\sum_i s_i C)$的
        - “二进制展开”：则可以大量减少重复。就是使用二进制数对si进行拆分。比如si为10，先取出1个，10-1=9，再取出2个，9-2=7，再取出4个，7-4=3，停止。即10拆为1，2，4，3，这4个物品，各物品的体积/价值分别为原来的1倍、2倍、4倍、3倍，数量为1.这样如果原来取1个物品，对应到取物品1，如果原来取2个物品，对应到取物品2，如果取3个，对应到取物品1和2，或者3，如果取4个，对应到取物品4，或者1和3，如果取5个，对应到取物品1和4或者2和3，如果取6个，对应到取物品2和4，如果取7个，对应到取物品1，2和4，或者3和4，如果取8个，对应到取物品1，3和4，如果取9个，对应到2，3和4，如果取10个，就是全取1，2，3和4.这样重复的情况大量减少。整体复杂度是$O(\sum_i \log s_i C)$的
    - 单调队列优化：为了解决多重背包中最内层的“决策”循环。使用滑动窗口求最值的思想。如果希望始终从队头取值更新的话，需要维持「队列元素单调」和「特定的窗口大小」。[细节参考](https://mp.weixin.qq.com/s?__biz=MzU4NDE3MTEyMA==&mid=2247486963&idx=1&sn=51d34f0f841122ed9be2335a402041e8&chksm=fd9ca6eccaeb2ffa1abe413177be376799b427b092bfb73c13e7b77e171b460f4c24b3b7d3bc&cur_album_id=1751702161341628417&scene=189#wechat_redirect)

三类传统背包问题的「一维空间优化」方式都已经讲过了。

我们发现，只有「完全背包」的「一维空间优化」是存在数学意义上的优化（能够有效降低算法时间复杂度）。

「01 背包」和「多重背包」的「一维空间优化」其实只是基于「朴素二维」解法做单纯的「滚动」操作而已（利用状态之间的依赖关系，配合遍历顺序，使得不再需要参与转移的空间能够被重新利用）。

#### 小结

- 01 背包：强调每件物品「只能选择一次」。对其进行「一维空间优化」并不能降低时间复杂度，进行「一维空间优化」时要求「容量维度“从大到小”进行遍历」。

- 完全背包：强调每件物品「可以选择无限次」。对其进行「一维空间优化」具有数学意义，可以将时间复杂度从$O(NC^2)$
降低到$O(NC)$，进行「一维空间优化」时要求「容量维度“从小到大”进行遍历」。

- 多重背包：强调每件物品「只能选择有限次」。对其无论是进行「一维空间优化」还是「普通扁平化」都不能降低时间复杂度，要应用额外的优化手段：二进制优化 或 单调队列优化 。

- 混合背包问题：背包里以上三种物品都有。背包问题的外层是可以沿着物品进行遍历的！
    1. 先将一个「多重背包」问题通过「二进制优化」的思路，转化为「01 背包」问题。
    2. 然后根据物品是属于「01 背包」还是「完全背包」决定容量是"从大到小"还是"从小到大"进行推算。换句话说就是根据物品的类型不同，选择不同的转移方式。
    
- 分组背包：把多个物品化为一组，要求每组中最多取一个。方法：最外层沿组进行遍历，每组内按照取哪个物品进行遍历。

同时前三种背包问题都有**不超过**和**恰好**两种状态定义。

这两种状态定义只在「初始化」上有区别。

至于该如何初始化则要抓住 什么样的状态是合法的 ：

- 对于「不超过」的状态定义：$f[0][x]$ 均为合法值0：代表不考虑任何物品，背包容量「不超过$x$」的所取得的最大价值为0

- 对于「恰好」的状态定义：$f[0][x]$中只有$f[0][0]$为合法值$0$，其他值均为“无效值”。代表不考虑任何物品，只有背包容量「恰好为0」时所取得的最大价值为0；其他容量「恰好为非0」是无法取得有效价值的（因为不考虑任何物品）。

````python
# 887 递归，楼层差值版，超时
# 超时用例 4, 5000
from collections import defaultdict
class Solution:
    def recurse(self, diff, k):
        if diff == 0: # 楼层差值为0时，直接返回
            self.memo[diff][k] = 0
            return
        if k == 1: # 鸡蛋个数为1的时候，需要从下向上遍历，次数为差值对应的次数
            self.memo[diff][1] = diff
            return
        cur_min = diff + 1
        for i in range(1, diff + 1): # 对尝试的楼层进行遍历
            # 当前层扔鸡蛋，鸡蛋破了的情况
            if i - 1 not in self.memo or k -1 not in self.memo[i - 1]: 
                self.recurse(i - 1, k - 1)
            lower_moves = self.memo[i - 1][k - 1]
            # 如果当前鸡蛋没有破
            higher_diff = diff - i
            if higher_diff not in self.memo or k not in self.memo[higher_diff]:
                self.recurse(higher_diff, k)
            higher_moves = self.memo[higher_diff][k]
            cur_min = min(cur_min, 1 + max(lower_moves, higher_moves))
        self.memo[diff][k] = cur_min
        return

    def superEggDrop(self, k: int, n: int) -> int:
        self.memo = defaultdict(dict)
        self.recurse(n, k) # 使用楼层差值做递归
        return self.memo[n][k]
    
````

````python
# 887 动规+二分搜索+空间优化
# time 9772 ms
# space 15.4MB
class Solution:
    def findMin(self, diff, egg): # 二分搜索
        l = 1
        r = diff
        while l < r:
            m = (l + r) // 2 + 1
            if self.memo[1-egg][m-1]>=self.memo[egg][diff - m]:
                r = m - 1
            else:
                l = m
        smaller_i = l
        return min(self.memo[egg][diff-smaller_i], self.memo[1-egg][smaller_i])
        
    def superEggDrop(self, k: int, n: int) -> int:
        self.memo = [[i for i in range(n+1)] for _ in range(2)]
        for i in range(2):
            self.memo[i][0] = 0
        for egg in range(1, k): # the egg number is egg + 1
            for diff in range(1, n + 1):
                cur_min = self.findMin(diff, egg % 2)
                self.memo[egg % 2][diff] = 1 + cur_min
        return self.memo[(k - 1) % 2][n]
````

区间问题的特点：
- 因为要枚举区间的开始和区间的结尾，所以所有的可能性有n^2种，即空间复杂度为n^2种。
- 一般是从区间长度为0或者1开始枚举，根据区间的长度进行遍历
- 状态的表示方法：区间的起始坐标+区间内要求的值
- 状态转移方程：大区间的值由子小区间组合而来，枚举子小区间的可能性


模板
1. initialize the subpart with the length of 0 or 1
2. iterate on the subpart length 
    3. iterate on the subpart start_id: get the value of matrix[start_id][start_id + len] based on the sub-subparts within it.

````python
# 375 dp
class Solution:
    def getMoneyAmount(self, n: int) -> int:
        allguess = [[-1 for i in range(1+n)] for _ in range(1+n)]
        # 1. 初始化区间长度为0的情况
        for i in range(1+n):
            allguess[i][i] = 0
        # 2. 对区间长度进行遍历
        for diff in range(1, n):
            # 对固定区间长度的区间起点进行遍历
            for start in range(1, 1 + n - diff):
                # 根据子区间的状态更新待求区间的状态
                min_guess = min(start + allguess[start + 1][start + diff], allguess[start][start + diff - 1] + start + diff)
                for k in range(start+1, start + diff):
                    min_guess = min(min_guess, k + max(allguess[start][k - 1], allguess[k+1][start + diff]))
                allguess[start][start + diff] = min_guess
        return allguess[1][n]
````

# 总结

- [X] 什么时候用递归？什么时候用动规？

从一个更高的层面来说，递归和动规都是在一个状态空间的图里做搜索，找到一条到达目标节点的路线。每一个状态可以看作图中的一个点。递归，通常来说，是从终点状态一步步逆着走回去，不一定走过所有的状态。而动规，则是选择一个起点状态，一步步走向终点，会遍历所有可能的状态。

总结而来，
- 如果终点是明确的，起点也是明确的，动规和递归的复杂度差不多，但是动规使用遍历，实现起来会简单很多，而且或许可以使用空间优化的方法。此时，选择动规更合适。例题740
- 如果终点是明确的，起点不确定，动规需要遍历所有的状态，而递归则从后往前，不一定要遍历所有的状态，此时，递归往往是更优的选择。例题87.

例题87，递归+记忆的时间是68ms，动规的时间是996ms。动规的访问次数显著多于递归。因为动规的起始状态包含了所有可能的状态，所以动规要走到终点，实际上遍历了每一个状态；而递归则是目标导向型的，从终点往回走，只搜索部分状态。这种情况下，递归是更优的。

例题740，递归+记忆的时间是336ms，动规的时间是44ms。在这道题中，两种方法实际搜索到的状态树是一样的，但是由于递归的调用需要额外开辟空间，整体操作不如遍历来的快，所以从实现的角度上来说，动规是更好的选择。

    - 如果当前状态的更新，可由相邻状态推导出来的话，可尝试进行空间压缩。沿相邻轴进行遍历，更新另一个维度的状态。

- [X] 为什么例题332沿两个方向压缩，都能得到正确的结果？例题518呢？
    - 332是巧了，min函数可以传递。真正正确的做法应该是沿coin维度进行遍历，更新amount的信息。
    - 518的sum函数，就不能传递。只能沿coin维度进行遍历。如果沿amount维度进行遍历，对每个amount，遍历到i的时候，更新的状态其实表示把coin[i]加入amount时候的组合数，综合来看得到的是排列数。
