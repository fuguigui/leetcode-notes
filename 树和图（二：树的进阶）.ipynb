- [树状数组](https://leetcode.cn/problems/range-sum-query-mutable/solution/shu-zhuang-shu-zu-xian-duan-shu-pythonsh-tjn5/)
    - 思想：以树的方式组织数组，用于快速计算区间和。奇数节点（下标的lowbit为1，下标从1开始计数）对应的是原数组中的值。下标的lowbit为n，就代表了从下标开始的前n个数的和（包括下标）。一个数的lowbit代表：其二进制表示中第一个为1的最低位对应的数，可用`x&-x`来计算。
    - 性质: 
        - 树的大小和数组的大小一样（为了方便计算下标，可以多一个出来，空出第一位）
        - `tree[i] = num[i] + num[i-1] + ... + num[i-n+1], n=lowbit(i)`
        - `tree[i] = num[i] + \sum_j tree[j], j: j+lowbit(j) = i`
        - `preSum[i] = tree[i] + tree[i2=i-lowbit(i)] + tree[i3=i2-lowbit(i2)] + ... (ik >=0)`，下标为n的前缀和为以n为下标的树的值，加上前一棵树的值，加上前前一棵树的值，直到没有更向前的树。下标为k的树的前一棵树下标的计算方式k-lowbit(k)
    - 模板：从1开始构造树的叶子节点，自下而上的构造。树的最后的节点，是最大的父节点。
        - 构造：用数组加一个前缀0，来初始化树。遍历数组，将它的值加到对应的上一级树节点上，假如当前的下标为i，则对应的上一级树节点的下标为`i+lowbit(i)`
        - 更新某个下标的值：
            1. 先求出这个下标原始的值，即区间为[i, i]的和
            2. 根据修改后的值，求出变化量，
            3. 然后向上传递，对于每一个父节点，都应用这个变化量。
        - 求给定区间的和：使用前缀和，求区间右端点对应的前缀和，和区间左端点的前一个端点对应的前缀和，两者相减即可。
        - 求前缀和：从该下标开始：依次找到前面的树的根，将值相加。
            - 什么是前面的树的根？就是去掉该节点和前面的节点相连的根之后，前面的节点所对应的最上端的根。
            - 如何求？下标为j的，前一个这样的根的下标是j-lowbit(j)，依次类推。
    - 复杂度：树状数组的更新和查询都是O(logn)的
        
- [线段树](https://books.halfrost.com/leetcode/ChapterThree/Segment_Tree/)：二叉树，其中每个节点代表一个区间。父节点代表的区间是子节点的并集。
     - [X] 如何表示区间？假设数据存在 size 为 n 的 arr[] 中。（如果不满足这样的假设呢？）
        - 线段树的根通常代表整个数据区间。这里是 arr[0：n-1]。
        - 树的每个叶子代表一个范围，其中仅包含一个元素。 因此，叶子代表 arr[0]，arr[1] 等等，直到 arr[n-1]。
        - 树的内部节点将代表其子节点的合并（求和）或并集结果。
        - 每个子节点可代表其父节点所代表范围的大约一半。(二分的思想)
     - 思想：使用数组，自上而下的构造树，叶子从左到右对应着原数组的值。
         - 线段树的根是在最前面，是**自上而下**展开的，这和树状数组不一样，
         - 树状数组是**自下而上**，根存储在数组的最后
     - 模板：
         - 构造：自上而下的构造：每个节点代表了一个给定连续区间，区间的左端点为l，右端点为r（均可取到）。
             - 当l=r的时候，找到了叶子节点，构造结束，将结果返回上一层
             - 当l<r的时候，递归构造两个叶子节点，拿到叶子节点的结果，“汇合”为本节点的结果
         - 更新某个下标的值：自下而上的更新，先更新叶子的值，再依次向上
         - 查询给定区间的“汇合”值：
             - 判断给定区间和当前区间的关系，选择是查询左叶子节点还是右叶子节点，还是“汇合”左右叶子节点的结果。
         - 这里的“汇合”：举例比如求和，取最大值等。要求是满足加法性质。

```python
# Trie Implementation
# 208 支持：插入，搜索，查前缀操作。删除功能为自己添加的
class TrieNode:
    def __init__(self, val=None):
        self.val = val
        self.end = 0 # 记录以当前字母为结尾的单词的个数
        self.path = 0 # 记录在特定位置经过该字母的单词个数
        self.child = {} # 这个字母的孩子节点
        if val:
            self.path = 1

    def addChild(self, val):
        self.path += 1 # 该字母的path加1
        if val in self.child:
            child_node = self.child[val]
            child_node.path += 1 # 如果已经有对应字母的child，将child的path加1
        else:
            child_node = TrieNode(val)
            self.child[val] = child_node

class Trie:
    def __init__(self):
        self.root = TrieNode()
    
    def insert(self, word: str) -> None:
        head = self.root
        for c in word:
            head.addChild(c)
            head = head.child[c]
        head.end += 1 # 结尾字母的end 加1

    def search(self, word: str) -> bool:
        head = self.root
        for c in word:
            if c not in head.child:
                return False
            head = head.child[c]
        return head.end > 0 # 判断是否有以该字母结尾的单词

    def startsWith(self, prefix: str) -> bool:
        head = self.root
        for c in prefix:
            if c not in head.child:
                return False
            head = head.child[c]
        return True
    
    def delete(self, word: str): # 增加了删除功能
        if not self.search(word): # 先查找是否字典树中是否有该单词。没有就不删除
            return
        head = self.root
        for c in word:
            child = head.child[c]
            if child.path == 1:
                head.child.pop(c)
                break
            else:
                child.path -= 1
                head = child
        head.end -= 1

# Your Trie object will be instantiated and called as such:
# obj = Trie()
# obj.insert(word)
# param_2 = obj.search(word)
# param_3 = obj.startsWith(prefix)
```

```python
# 树状数组模板 https://leetcode.cn/problems/range-sum-query-mutable/solution/shu-zhuang-shu-zu-xian-duan-shu-pythonsh-tjn5/

'''BIT：Binary Indexed Tree 树状数组'''
class NumArray:
    
    '''自定义lowbit函数'''
    def lowbit(self, x: int) -> int:
        return x&(-x)

    def __init__(self, nums: List[int]):
        self.tree = [0] + nums              # 构造出的BIT空间比原nums多一，第1个下标0不用
        for i in range(1, len(self.tree)):  # 这种方式构造的BIT时间复杂度为O(n)
            j = i + self.lowbit(i)          # 将当前树的值加到对应的上一级的树节点上
            if j < len(self.tree):
                self.tree[j] += self.tree[i]

    def update(self, index: int, val: int) -> None:
        pre_val = self.sumRange(index, index)   # 先求出这个下标原始的值，即区间为[i, i]的和
        delta = val - pre_val           # 变更值
        i = index + 1                   # i: 对应数值在BIT中的位置（index+1）
        while i < len(self.tree): # 向上传递，对于每一个父节点，都应用这个变化量。
            self.tree[i] += delta
            i += self.lowbit(i) # 寻找再上一层的父节点

    
    def sumRange(self, left: int, right: int) -> int:
        return self.preSum(right) - self.preSum(left-1)   # right的前缀和 减去 left-1的前缀和

    '''自定义前缀和preSum函数'''
    def preSum(self, index: int) -> int:    # index：在原nums中的位置
        i = index + 1                       # i: 对应数值在BIT中的位置（index+1）
        summ = 0
        while i: # 从该下标开始：依次找到前面的树的根，将值相加。
            summ += self.tree[i]
            i -= self.lowbit(i) # 求前面的树的根对应的下标
            # i &= i-1        # 跟上一行等价
        return summ
    

```

```python
# 线段树模板

class NumArray:

    def __init__(self, nums: List[int]):
        self.len = len(nums)
        self.tree = [None] * 4 * self.len
        self.nums = nums
        self.nid2tid = [None] * self.len # 存储数组的下标和树中叶子节点下标的对应关系
        self.build(0, 0, self.len-1) # 自上而下的构造
    
    def build(self, i, l, r):
        if l == r: # 当l=r的时候，找到了叶子节点，构造结束，将结果返回上一层
            self.tree[i] = self.nums[l]
            self.nid2tid[l] = i
            return 
        mid = (l + r ) // 2
        # 当l<r的时候，递归构造两个叶子节点，拿到叶子节点的结果，“汇合”为本节点的结果
        self.build(2 * i + 1, l, mid)
        self.build(2 * i + 2, mid + 1, r)
        self.tree[i] = self.tree[2*i+1] + self.tree[2*i+2]

    def query(self, tid, left, right, tleft, tright):
        if left == tleft and right == tright: # 恰好是本节点负责的范围，直接返回
            return self.tree[tid]
        tmid = (tleft + tright) // 2
        if right <= tmid:
            return self.query(2*tid + 1, left, right, tleft, tmid)  # 是左子树负责的范围，递归去左子树查询
        if left >= tmid + 1:
            return self.query(2*tid + 2, left, right, tmid + 1, tright) # 是右子树负责的范围，递归去右子树查询
        # 左右各负责一部分，分别查询，再汇总
        return self.query(2*tid + 1, left, tmid, tleft, tmid) + self.query(2*tid + 2, tmid + 1, right, tmid + 1, tright)

    def update(self, index: int, val: int) -> None:
        diff = val - self.nums[index]
        nid = self.nid2tid[index] # 找到叶子节点的下标，
        while nid >= 0: # 自下向上地更新
            self.tree[nid] += diff
            nid = (nid + 1) // 2 - 1
        self.nums[index] = val # 不要忘记更新数组中的值，用于方便计算delta
        
    def sumRange(self, left: int, right: int) -> int:
        return self.query(0, left, right, 0, self.len-1)
```

```python
class WordNode:
    def __init__(self):
        self.end = False
        self.children = {}

class WordDictionary:
    def __init__(self):
        self.root = WordNode()

    def addWord(self, word: str) -> None:
        head = self.root
        for l in word:
            if l not in head.children:
                head.children[l] = WordNode()
            head = head.children[l]
        head.end = True

    def bfsSearch(self, word):
        next_layer = [self.root]
        for l in word:
            cur_layer = next_layer
            next_layer = []
            if l == '.':
                for node in cur_layer:
                    for key, child in node.children.items():
                        next_layer.append(child)
            else:
                for node in cur_layer:
                    if l in node.children:
                        next_layer.append(node.children[l])
            if not next_layer:
                return False
        for node in next_layer:
            if node.end:
                return True
        return False

    def dfsSearch(self, word, node):
        last_l = word[0]
        if len(word) == 1:
            if last_l == '.':
                for key, child in node.children.items():
                    if child.end > 0:
                        return True
            elif last_l in node.children and node.children[last_l].end > 0:
                return True
            return False
        if last_l == '.':
            for key, child in node.children.items():
                if self.dfsSearch(word[1:], child):
                    return True
        elif last_l in node.children:
            return self.dfsSearch(word[1:], node.children[last_l])
        return False

    def search(self, word: str) -> bool:
        if len(word) > 10:
            return self.dfsSearch(word, self.root)
        return self.bfsSearch(word)



# Your WordDictionary object will be instantiated and called as such:
# obj = WordDictionary()
# obj.addWord(word)
# param_2 = obj.search(word)
```

```python
# 307 思路一：直接求和，超时
class NumArray:

    def __init__(self, nums: List[int]):
        self.nums = nums

    def update(self, index: int, val: int) -> None:
        self.nums[index] = val

    def sumRange(self, left: int, right: int) -> int:
        return sum(self.nums[left:right + 1])
    
# 思路一的优化版
class NumArray:

    def __init__(self, nums: List[int]):
        self.nums = nums 
        self.total = sum(self.nums)
        
    def update(self, index: int, val: int) -> None:
        pre_val, self.nums[index] = self.nums[index], val
        self.total = self.total - pre_val + val
        
    def sumRange(self, left: int, right: int) -> int:
        return sum(self.nums[left:right+1]) if right-left < len(self.nums)//2 else self.total - sum(self.nums[:left]) - sum(self.nums[right+1:])
    
# 思路二：前缀和，超时
class NumArray:
    def __init__(self, nums: List[int]):
        self.nums = nums
        self.pre_sum = [0]
        for i in nums:
            self.pre_sum.append(self.pre_sum[-1] + i)

    def update(self, index: int, val: int) -> None:
        diff = val - self.nums[index]
        self.nums[index] = val
        for i in range(index + 1, len(self.nums) + 1):
            self.pre_sum[i] += diff

    def sumRange(self, left: int, right: int) -> int:
        return self.pre_sum[right + 1] - self.pre_sum[left]

# 思路三：树状数组
'''BIT：Binary Indexed Tree 树状数组'''
class NumArray:
    
    '''自定义lowbit函数'''
    def lowbit(self, x: int) -> int:
        return x&(-x)

    def __init__(self, nums: List[int]):
        self.tree = [0] + nums              # 构造出的BIT空间比原nums多一，第1个下标0不用
        for i in range(1, len(self.tree)):  # 这种方式构造的BIT时间复杂度为O(n)
            j = i + self.lowbit(i)          # 构造BIT的巧妙方式
            if j < len(self.tree):
                self.tree[j] += self.tree[i]

    def update(self, index: int, val: int) -> None:
        pre_val = self.sumRange(index, index)   # index：在原nums中的位置
        delta = val - pre_val           # 变更值
        i = index + 1                   # i: 对应数值在BIT中的位置（index+1）
        while i < len(self.tree):
            self.tree[i] += delta
            i += self.lowbit(i)

    '''自定义前缀和preSum函数'''
    def preSum(self, index: int) -> int:    # index：在原nums中的位置
        i = index + 1                       # i: 对应数值在BIT中的位置（index+1）
        summ = 0
        while i:
            summ += self.tree[i]
            i -= self.lowbit(i)
            # i &= i-1        # 跟上一行等价
        return summ
    
    def sumRange(self, left: int, right: int) -> int:
        return self.preSum(right) - self.preSum(left-1)   # right的前缀和 减去 left-1的前缀和



# Your NumArray object will be instantiated and called as such:
# obj = NumArray(nums)
# obj.update(index,val)
# param_2 = obj.sumRange(left,right)

# 思路四：线段树
class NumArray:

    def __init__(self, nums: List[int]):
        self.len = len(nums)
        self.tree = [None] * 4 * self.len
        self.nums = nums
        self.nid2tid = [None] * self.len
        self.build(0, 0, self.len-1)
        # print(self.tree)
        # print(self.nid2tid)
    
    def build(self, i, l, r):
        if l == r:
            self.tree[i] = self.nums[l]
            self.nid2tid[l] = i
            return 
        mid = (l + r ) // 2
        self.build(2 * i + 1, l, mid)
        self.build(2 * i + 2, mid + 1, r)
        self.tree[i] = self.tree[2*i+1] + self.tree[2*i+2]

    def query(self, tid, left, right, tleft, tright):
        # print("Querying: ", left, right, "for the treeid: ", tid, " within: ", tleft, tright)
        if left == tleft and right == tright:
            return self.tree[tid]
        tmid = (tleft + tright) // 2
        if right <= tmid:
            return self.query(2*tid + 1, left, right, tleft, tmid)
        if left >= tmid + 1:
            return self.query(2*tid + 2, left, right, tmid + 1, tright)
        return self.query(2*tid + 1, left, tmid, tleft, tmid) + self.query(2*tid + 2, tmid + 1, right, tmid + 1, tright)

    def update(self, index: int, val: int) -> None:
        diff = val - self.nums[index]
        nid = self.nid2tid[index]
        while nid >= 0:
            self.tree[nid] += diff
            nid = (nid + 1) // 2 - 1
        # print("Tree after updating: ", index, self.nums[index], val, self.tree)
        self.nums[index] = val
        
    def sumRange(self, left: int, right: int) -> int:
        return self.query(0, left, right, 0, self.len-1)
        


# Your NumArray object will be instantiated and called as such:
# obj = NumArray(nums)
# obj.update(index,val)
# param_2 = obj.sumRange(left,right)
```

```python
# 315 思路一：双层循环
class Solution:
    def countSmaller(self, nums: List[int]) -> List[int]:
        res = [] 
        for i, num in enumerate(nums):
            cnt = 0
            for j in range(i+1, len(nums)):
                if nums[j] < num:
                    cnt += 1
            res.append(cnt)
        return res
    
# 思路二：逆序遍历+二分查找+排序 time 3012ms, space 31.1MB
class Solution:
    def __init__(self):
        self.sorted = []

    def getCurID(self, num):
        l = 0
        r = len(self.sorted) - 1
        if num > self.sorted[-1]:
            return r + 1
        if num <= self.sorted[0]:
            return l
        while l < r:
            mid = (l+r)//2
            if self.sorted[mid] >= num:
                r = mid
            else:
                l = mid + 1
        return l

    def findSmallerNumber(self, num):
        if not self.sorted:
            self.sorted.append(num)
            return 0
        cur_id = self.getCurID(num)
        self.sorted.insert(cur_id, num)
        return cur_id

    def countSmaller(self, nums: List[int]) -> List[int]:
        res = []
        ttl = len(nums)
        for i in range(ttl - 1, -1, -1):
            smaller_num = self.findSmallerNumber(nums[i])
            res.append(smaller_num)
        return res[::-1]
    
# 思路二：python的数据结构版本
from sortedcontainers import SortedList
class Solution:
    def countSmaller(self, nums: List[int]) -> List[int]:

        n = len(nums)
        res = [0] * n
        sl = SortedList()

        for i in range(n-1, -1, -1):        # 反向遍历
            cnt = sl.bisect_left(nums[i])   # 找到右边比当前值小的元素个数
            res[i] = cnt                    # 记入答案
            sl.add(nums[i])                 # 将当前值加入有序数组中
        
        return res
    
# 思路四：树状数组 time 1712ms, space 33.4MB
class Solution:
    def __init__(self):
        self.tree = []
        self.n = 0

    def lowbit(self, x):
        return x & (-x)

    def buildTree(self, n):
        self.n = n + 1
        self.tree = [0] * (n + 1)

    def query(self, i):
        pre_sum = 0
        tid = i + 1 # 注意要加1
        while tid > 0: # 注意是大于0，而不是大于等于0
            pre_sum += self.tree[tid]
            tid -= self.lowbit(tid)
        return pre_sum

    def add(self, i, val):
        tid = i + 1 # 注意要加1
        while tid < self.n:
            self.tree[tid] += val
            tid += self.lowbit(tid)

    def countSmaller(self, nums: List[int]) -> List[int]:
        res = []
        uniques = sorted(set(nums))
        ranks = {num: rank for rank, num in enumerate(uniques)} # 小trick：离散化
        self.buildTree(len(uniques))
        for i in range(len(nums) -1, -1, -1):
            ranki = ranks[nums[i]]
            res.append(self.query(ranki - 1))
            self.add(ranki, 1)
        return res[::-1]
    
# 思路五：线段树 time 3248ms, space 35.4MB
class Solution:
    def __init__(self):
        self.tree = []
        self.r = 0
        self.nid2tid = {}


    def buildTree(self, n):
        self.tree = [0] * 4 * n
        self.r = n - 1
        self.recurBuild(0, 0, n - 1)

    def recurBuild(self, tid, l, r):
        if l == r:
            self.nid2tid[l] = tid
            return
        m = (l + r) // 2
        self.recurBuild(2*tid + 1, l, m)
        self.recurBuild(2*tid + 2, m + 1, r)

    def query(self, r):
        if r < 0:
            return 0
        return self.queryInTree(0, 0, self.r, 0, r)
        
    def queryInTree(self, tid, tleft, tright, qleft, qright):
        if tleft == qleft and tright == qright:
            return self.tree[tid]
        tmid = (tleft + tright) // 2
        if qleft >= tmid + 1:
            return self.queryInTree(2*tid + 2, tmid + 1, tright, qleft, qright)
        if qright <= tmid:
            return self.queryInTree(2*tid+1, tleft, tmid, qleft, qright)
        return self.queryInTree(2*tid+1, tleft, tmid, qleft, tmid) + self.queryInTree(2*tid + 2, tmid + 1, tright, tmid + 1, qright)

    def add(self, i, val):
        tid = self.nid2tid[i]
        while tid >= 0:
            self.tree[tid] += val
            tid = (tid + 1) // 2 - 1

    def countSmaller(self, nums: List[int]) -> List[int]:
        res = []
        uniques = sorted(set(nums))
        ranks = {num: rank for rank, num in enumerate(uniques)}
        self.buildTree(len(uniques))
        for i in range(len(nums) -1, -1, -1):
            ranki = ranks[nums[i]]
            res.append(self.query(ranki - 1))
            self.add(ranki, 1)
        return res[::-1]
    
# 思路五：题解版 time 1896ms, space 32.2MB
''''ST：Segment Tree 线段树'''
class ST:
    def __init__(self, n):
        self.n = n
        self.tree = [0] * (2*n)     # 元素数目为原数组的2倍
    
    def add(self, i, delta):
        i += self.n                 # 原数组下标转换到线段树下标
        while i>0:
            self.tree[i] += delta
            i //= 2

    def rangeSum(self, i, j):       # 区间和【自下而上统计】
        i, j = i+self.n, j+self.n   # 原数组下标转换到线段树下标
        summ = 0
        while i<=j:
            if i&1 == 1:        # 右子节点
                summ += self.tree[i]
                i += 1
            if j&1 == 0:        # 左子节点
                summ += self.tree[j]
                j -= 1
            i //= 2
            j //= 2
        return summ


class Solution:
    def countSmaller(self, nums: List[int]) -> List[int]:

        n = len(nums)
        
        # 离散化：绝对数值转秩次【rank从0开始】
        uniques = sorted(set(nums))
        rank_map = {v:i for i,v in enumerate(uniques)}
        
        # 构建线段树
        tree = ST(len(uniques))
        
        # 从右往左查询
        res = [0] * n
        for i in range(n-1, -1, -1):
            rank = rank_map[nums[i]]    # 当前值的排名
            tree.add(rank, 1)           # 单点更新+1
            res[i] = tree.rangeSum(0, rank-1)    # 查询 当前排名之前 的元素有多少个

        return res
```

# 总结

来自宫水三叶：

针对不同的题目，我们有不同的方案可以选择（假设我们有一个数组）：

    数组不变，求区间和：「前缀和」、「树状数组」、「线段树」
    多次修改某个数（单点），求区间和：「树状数组」、「线段树」
    多次修改某个区间，输出最终结果：「差分」
    多次修改某个区间，求区间和：「线段树」、「树状数组」（看修改区间范围大小）
    多次将某个区间变成同一个数，求区间和：「线段树」、「树状数组」（看修改区间范围大小）

这样看来，「线段树」能解决的问题是最多的，那我们是不是无论什么情况都写「线段树」呢？

答案并不是，而且恰好相反，只有在我们遇到第 4 类问题，不得不写「线段树」的时候，我们才考虑线段树。

因为「线段树」代码很长，而且常数很大，实际表现不算很好。我们只有在不得不用的时候才考虑「线段树」。

总结一下，我们应该按这样的优先级进行考虑：

- 简单求区间和，用「前缀和」
- 多次将某个区间变成同一个数，用「线段树」
- 其他情况，用「树状数组」
