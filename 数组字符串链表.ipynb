# 理论知识

数据结构的存储方式只有两种：链表（链式存储）和数组（顺序存储）
数据结构的基本操作：遍历+访问，更具体的为增删查改

- 遍历和访问就两种基本形式：线性（迭代）和非线性（递归）的
    - 线性：数组遍历，典型的迭代结构，for/while
    - 非线性：递归结构。链表可以线性也可以非线性。
        - 链表遍历，到二叉树遍历，到N叉树遍历，到图的遍历（使用visited数组记录状态）

- 链表：优势：增删
- 数组：优势：查

## 链表

在进行链表**删除**操作的小技巧：
经常会因为对当前节点进行操作而导致内存或 指针出现问题。有两个小技巧可以解决这个问题:
- 一是尽量处理当前节点的下一个节点而非当前 节点本身，
- 二是建立一个虚拟节点 (dummy node)，使其指向当前链表的头节点，这样即使原链表所有节点全被删除，也会有一个 dummy存在，返回 dummy->next 即可。


主要思考何时进行操作。是在当前节点对下一个进行操作？还是记录pre节点，对当前节点进行操作？
使用迭代还是递归？递归会需要多余的栈空间。


### 常用的技巧

- 虚拟头节点：如果头节点可能会发生变化/链表可能为空，那么就设置一个虚拟头节点，指向真正的头节点。能减少很多单独的分类讨论。例如21合并2个有序链表（链表可能为空），19删除倒数第n个（头节点可能被删除）

- 快慢指针：用途
    - 在不知道长度的链表中，进行计数，例题19
    - 寻找到链表中具有特定特征的位置，例题141, 160
    - 判断链表是否有环：快慢指针是否会重合；进阶：寻找环的起点：慢指针回到起点，再次相遇的地方。141，142，160

- 前序遍历/后序遍历：链表兼具递归结构，也可以有前序遍历和后序遍历

```
void traverse(ListNode head) {
    // 前序遍历代码
    traverse(head.next);
    // 后序遍历代码
}

```

### 反转链表

从易到难可依次分为：反转整个链表（206），反转链表的前n个数，反转链表的任意指定片段（92）。既可以用递归又可以用迭代。但是递归操作链表并不高效，和迭代解法相比，虽然时间复杂度都是O(N)，但是迭代解法的空间复杂度是O(1)，而递归解法需要堆栈，空间复杂度是O(N)。所以，考虑效率的话，还是迭代叭～

- [(easy)206 reverse linked list](https://leetcode.cn/problems/reverse-linked-list/)
    - 思路一：递归。对下一个节点进行reverse，再把head放到reverse后的最后一位。注意处理边缘情况。空间复杂度O(n).空间复杂度主要取决于递归调用的栈空间，最多为 n 层。
    - 思路二：迭代。我的实现是不停地创建新的节点。空间复杂度O(n)
    - 思路二：迭代，不创建新的节点，每次先记录当前元素的下一个元素，然后把当前元素的下一个元素设为旧的头元素，再把当前元素设为新的头元素。空间复杂度O(1)
    
    递归和迭代的思考角度的区别：
    - 递归：思考的是：假如我之后的链表已经被翻转过来了，我该怎么做？放到翻转后的链表的尾部，并且返回新的链表。
    - 迭代：思考的是：假如我之前的链表已经被翻转过来了，我该怎么做？下一次应该从我的下一个开始，我该连到翻转过的链表的头部，并且作为新的头。
    
- [(medium)92 Reverse Linked List II](https://leetcode.cn/problems/reverse-linked-list-ii)
    - 思路一：递归：先找到要反转的起始位置的前一个位置，因为要更改该位置的后一个值。接着做对链表反转指定前n个数的操作。最后把反转后的链表连接到前一个位置的后面。
    - 思路二：迭代：找到需要开始反转的起始位置的前一个位置pre，在需要反转的区间里，维护一个尾元素tail，它的next指向当前需要反转的元素，把tail的next更新为要反转的元素的next，把要反转的元素的next更新为pre的next，将要反转的元素更新为pre的next。

````python
# 206 递归
# 执行用时：40 ms, 在所有 Python3 提交中击败了67.95% 的用户
# 内存消耗：20.7 MB, 在所有 Python3 提交中击败了14.85% 的用户
class Solution:
    def recurse(self, head):
        if not head or not head.next:
            return head, head
        reversed_head, reversed_tail = self.recurse(head.next)
        reversed_tail.next = ListNode(head.val)
        return reversed_head, reversed_tail.next

    def reverseList(self, head: ListNode) -> ListNode:
        reversed_head, _ = self.recurse(head)
        return reversed_head
    
# 迭代1
# 执行用时：44 ms, 在所有 Python3 提交中击败了41.20% 的用户
# 内存消耗：16.9 MB, 在所有 Python3 提交中击败了23.42% 的用户

class Solution:
    def reverseList(self, head: ListNode) -> ListNode:
        dummy_head = ListNode(next=head)
        old_head = None
        cur_node = dummy_head
        while cur_node.next:
            cur_node = cur_node.next
            new_head = ListNode(cur_node.val) # 这一步实际上创建了新的节点。
            new_head.next = old_head
            old_head = new_head
        return old_head
    
# 迭代2，不创建新节点
# 执行用时：44 ms, 在所有 Python3 提交中击败了41.20% 的用户
# 内存消耗：16 MB, 在所有 Python3 提交中击败了73.84% 的用户
class Solution:
    def reverseList(self, head: ListNode) -> ListNode:
        new_head = None
        cur_next = head
        while cur_next:
            new_next = cur_next.next # 下一个当前元素
            cur_next.next = new_head # 把当前元素的下一个设为旧的头
            new_head = cur_next # 当前元素作为新的头
            cur_next = new_next # 更新当前元素
        return new_head
````

````python
# 23 迭代
# 执行用时：88 ms, 在所有 Python3 提交中击败了46.49% 的用户
# 内存消耗：18.2 MB, 在所有 Python3 提交中击败了42.48% 的用户
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
import heapq
class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        heap = []
        for i, sub_list in enumerate(lists):
            if sub_list:
                heapq.heappush(heap, (sub_list.val, i))
        head = ListNode(0)
        tail = head
        while heap:
            _, top_id = heapq.heappop(heap)
            cur_node = lists[top_id]
            tail.next = cur_node
            tail = tail.next
            if cur_node.next:
                cur_node = cur_node.next
                lists[top_id] = cur_node
                heapq.heappush(heap, (cur_node.val, top_id))

        return head.next

# 递归
# 执行用时：76 ms, 在所有 Python3 提交中击败了68.94% 的用户
# 内存消耗：28 MB, 在所有 Python3 提交中击败了5.08% 的用户
import heapq
class Solution:
    def recurse(self, heap, lists):
        if not heap:
            return None
        top_val, top_i = heapq.heappop(heap)
        head = lists[top_i]
        lists[top_i] = head.next
        if head.next:
            heapq.heappush(heap, (head.next.val, top_i))
        head.next = self.recurse(heap, lists)
        return head

    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:
        heap = []
        for i, head in enumerate(lists):
            if head:
                heapq.heappush(heap, (head.val, i))
        return self.recurse(heap, lists)
````

````python
# 92 思路：递归
# 执行用时：32 ms, 在所有 Python3 提交中击败了91.22% 的用户
# 内存消耗：15.2 MB, 在所有 Python3 提交中击败了15.88% 的用户

class Solution:
    def recurse(self, head, num):
        if num == 1:
            return head, head
        next_head, next_tail = self.recurse(head.next, num - 1)
        head.next = next_tail.next
        next_tail.next = head
        return next_head, head

    def reverseBetween(self, head: ListNode, left: int, right: int) -> ListNode:
        dummy_head = ListNode(next=head)
        pre = dummy_head
        start_id = 1
        while start_id < left:
            pre = pre.next
            start_id += 1
        reversed_head, reversed_tail = self.recurse(pre.next, right - left + 1)
        pre.next = reversed_head
        return dummy_head.next

# 思路二：迭代
# 执行用时：40 ms, 在所有 Python3 提交中击败了47.52% 的用户
# 内存消耗：15.2 MB, 在所有 Python3 提交中击败了14.99% 的用户
class Solution:
    def reverseBetween(self, head: ListNode, left: int, right: int) -> ListNode:
        dummy_head = ListNode(next=head)
        pre = dummy_head
        start_id = 1
        while start_id < left:
            pre = pre.next
            start_id += 1

        tail = pre.next
        while start_id < right:
            cur_next = tail.next
            tail.next = cur_next.next
            cur_next.next = pre.next
            pre.next = cur_next
            start_id += 1
        return dummy_head.next
````

````python
# 25 思路一：整体递归，局部迭代
# 执行用时：44 ms, 在所有 Python3 提交中击败了87.17% 的用户
# 内存消耗：15.9 MB, 在所有 Python3 提交中击败了92.57% 的用户

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def reverse(self, head, k):
        cnt = 1
        old_head = head
        cur_node = head
        while cnt < k:
            new_head = cur_node.next
            cur_node.next = cur_node.next.next
            new_head.next = old_head
            old_head = new_head
            cnt +=1
        return old_head, cur_node
    
    def reverseKGroup(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
        p = head
        for i in range(k):
            if not p:
                return head
            p = p.next
        next_head = p
        # reverse from head until k nodes
        old_head, cur_node = self.reverse(head, k)
        if next_head:
            cur_node.next = self.reverseKGroup(next_head, k)
        return old_head
    
# 思路二：整体递归+局部递归
# 执行用时：44 ms, 在所有 Python3 提交中击败了87.17% 的用户
# 内存消耗：16.5 MB, 在所有 Python3 提交中击败了7.39% 的用户
class Solution:
    def reverse(self, head, k):
        if k <= 1:
            return head, head
        sub_head, sub_tail = self.reverse(head.next, k-1)
        head.next = sub_tail.next
        sub_tail.next = head
        return sub_head, head
    
    def reverseKGroup(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
        p = head
        for i in range(k):
            if not p:
                return head
            p = p.next
        next_head = p
        # reverse from head until k nodes
        old_head, cur_node = self.reverse(head, k)
        if next_head:
            cur_node.next = self.reverseKGroup(next_head, k)
        return old_head
````

````python
# 234 空间复杂度为O(1)的解
# 执行用时：672 ms, 在所有 Python3 提交中击败了65.04% 的用户
# 内存消耗：40.5 MB, 在所有 Python3 提交中击败了73.61% 的用户
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def findMiddle(self, head):
        faster = head
        slower = head
        while True:
            if not faster or not faster.next:
                break
            faster = faster.next.next
            slower = slower.next
        return slower
    
    def reverse(self, head):
        tail = head
        old_head = tail
        while tail.next:
            new_head = tail.next
            tail.next  = tail.next.next
            new_head.next = old_head
            old_head = new_head
        return old_head

    def isPalindrome(self, head: ListNode) -> bool:
        middle = self.findMiddle(head)
        second_head = self.reverse(middle)
        p1 = head
        p2 = second_head
        while p2:
            if p1.val != p2.val:
                return False
            p1 = p1.next
            p2 = p2.next
        return True
````

# 经验总结

    巧妙的构造虚拟头结点。可以使遍历处理逻辑更加统一。
    灵活使用递归。构造递归条件，使用递归可以巧妙的解题。不过需要注意有些题目不能使用递归，因为递归深度太深会导致超时和栈溢出。
    链表区间逆序。第 92 题。
    链表寻找中间节点。第 876 题。链表寻找倒数第 n 个节点。第 19 题。只需要一次遍历就可以得到答案。
    合并 K 个有序链表。第 21 题，第 23 题。
    链表归类。第 86 题，第 328 题。
    链表排序，时间复杂度要求 O(n * log n)，空间复杂度 O(1)。只有一种做法，归并排序，至顶向下归并。第 148 题。
    判断链表是否存在环，如果有环，输出环的交叉点的下标；判断 2 个链表是否有交叉点，如果有交叉点，输出交叉点。第 141 题，第 142 题，第 160 题。
