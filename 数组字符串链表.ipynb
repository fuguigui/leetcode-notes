````python
# 234 ç©ºé—´å¤æ‚åº¦ä¸ºO(1)çš„è§£
# æ‰§è¡Œç”¨æ—¶ï¼š672 ms, åœ¨æ‰€æœ‰ Python3 æäº¤ä¸­å‡»è´¥äº†65.04% çš„ç”¨æˆ·
# å†…å­˜æ¶ˆè€—ï¼š40.5 MB, åœ¨æ‰€æœ‰ Python3 æäº¤ä¸­å‡»è´¥äº†73.61% çš„ç”¨æˆ·
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def findMiddle(self, head):
        faster = head
        slower = head
        while True:
            if not faster or not faster.next:
                break
            faster = faster.next.next
            slower = slower.next
        return slower
    
    def reverse(self, head):
        tail = head
        old_head = tail
        while tail.next:
            new_head = tail.next
            tail.next  = tail.next.next
            new_head.next = old_head
            old_head = new_head
        return old_head

    def isPalindrome(self, head: ListNode) -> bool:
        middle = self.findMiddle(head)
        second_head = self.reverse(middle)
        p1 = head
        p2 = second_head
        while p2:
            if p1.val != p2.val:
                return False
            p1 = p1.next
            p2 = p2.next
        return True
````
