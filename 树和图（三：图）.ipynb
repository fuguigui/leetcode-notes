## 图上的算法

### 广搜模板

1. 任意选择一个点作为起点，计入当前层
2. 标记当前层的所有点为访问过，对于当前层的每个点，去看它的邻居们。把没有尚未访问的邻居加入下一层。
3. 如果下一层不为空，则把下一层记为当前层，重复第一步。
4. 如果下一层为空，根据题目，看看是另取新的点，还是结束搜索。


双向BFS模板


Dijkstra算法可以视为使用贪心策略优化后的广度优先搜索


### 深搜模板

1. 任意选择一个点作为起点
2. 如果这个起点没有尚未访问过的邻居，结束深搜
3. 从这个起点的尚未访问过的邻居中，选择一个，执行第1步，直到结束；选择下一个尚未访问过的邻居，直到没有邻居。

### 拓扑排序

其实是广搜的一种特殊变形，对应于广搜的每一步，拓扑排序的变形版本为
1. 选择所有入度为0的点作为起点，计入当前层
2. 对于当前层的每个点，去看它的邻居们（指向的节点），把邻居们的入度各减一，如果把某个邻居的入度降为0，将其加入下一层。
3. 如果下一层不为空，则把下一层记为当前层，重复第一步。
4. 如果下一层为空：如果存在点的入度不为0，则失败；否则，成功，根据要求返回。

```python
# 785 多次BFS
class Solution:
    def __init__(self):
        self.color = []

    def getUncolored(self):
        res = []
        for i, c in enumerate(self.color):
            if c == 0:
                res.append(i)
        return res
    
    def coloring(self, start_node, color):
        next2color = set([start_node])
        while next2color:
            cur2color = next2color
            next2color = set()
            for node in cur2color:
                self.color[node] = color
                for neigh in self.graph[node]:
                    if self.color[neigh] == 0:
                        next2color.add(neigh)
                    elif self.color[neigh] == color:
                        return False
            color = 3 - color
        return True

    def isBipartite(self, graph: List[List[int]]) -> bool:
        self.color = [0] * len(graph)
        self.graph = graph
        while True:
            uncolored_nodes = self.getUncolored() # 这个是开启多次染色过程
            if not uncolored_nodes:
                return True
            start_node = uncolored_nodes[0]
            if not self.coloring(start_node, 1): # 本次染色是否成功
                return False
        return False
```

## 遍历


- [(hard)127Word Ladder](https://leetcode.cn/problems/word-ladder/)
    - 思考以下几个问题
        - [X] 单词的长度如何影响算法的选择？设为`L`，线性影响时间复杂度
        - [X] 每个位置可变化的候选多少如何影响算法的选择？设为`C`
        - [X] 字典的大小如何影响算法的选择？设为`M`
        - [X] 什么时候确定一个点的子节点？有两种方式：一种是预先建图，另一种是遍历的过程中寻找子节点？有什么区别？
            - 先建图的话：是构建一个hashmap，key为各个word，value为与该word的相差一个字母的其他word。时间复杂度是和字典大小`M`呈线性关系的。空间复杂度：至多需要额外O(`M^2`)的空间来存储邻接关系。
            - 在遍历过程中寻找子节点的话，对遍历中的每一个word，寻找子节点。时间复杂度是和遍历次数`R`呈线性关系的。遍历次数总是小于等于`M`的。
            - 综上，选择后者
        - [X] 如何确定一个点的子节点？也有两种方式。
            - 一种是遍历字典中的所有word，判断该word与目标word是不是相差一个字母，时间复杂度为O(`ML`)的。遍历过程有`M`个，判断过程要判断`L`个位置。
            - 另一种是枚举与这个word相差一个字母的所有word，判断是不是在字典中，时间复杂度固定为O(`LC`)的。枚举的个数有`LC`个，判断是O(1)的。
            - 这么分析的话，如果字典的大小`M`小于每个位置可变化的候选个数`C`，就选择先遍历后判断，否则就选择先枚举后判断。
        
    - 初步想一下：有哪些方法可以解决这个问题？
        - DFS：不合适，这里是找最短路，DFS可能做很多无用功。
            1. 对于当前的根，找到所有的未被访问的子节点，
                - 如果子节点中有目标节点，就结束并返回
                - 否则，将所有的候选子节点标记为已访问
            2. 遍历所有候选子节点，进行下一步深搜。这一步有无选择顺序可言？
            3. 如果当前的根，没有未被访问的子节点，则返回失败。
            时间复杂度：对于每个节点可能会访问很多遍，时间复杂度比较高，最多可能会被访问到M遍，即最多O(`M^2`)次遍历。复杂度比BFS多了O(`M`)出来。空间复杂度：栈调用最多可能是O(`M`)深的。
            
        - BFS：
            1. 遍历当前层中的每一个点 （最多是O(M)次遍历）
            2. 找到该点的未被访问的子节点，这一步的复杂度如果是在遍历过程中寻找的话，
                - 如果该子节点是目标，则结束
                - 否则，将子节点标记为已访问，把子节点放入候选
            3. 将所有候选作为当前层，重复第一步
            - 时间复杂度：对于所有的点，最多会被遍历到一遍，对于每个遍历到的元素，找到未被访问的子节点的复杂度是O(`ML`)或O(`LC`)的，总体而言，复杂度是O(`M^2L`)或O(`MLC`)的，其中`C`可以认为是常数的，所以时间复杂度是O(`ML`)的。
            - 空间复杂度：记录是否visited是O(M)的，每次所有可行的子节点的长度最多是O(M)的，其他可认为是常数的。整体是O(M)的。
            
        - 双向BFS
            1. 从两个集合分别进行BFS，每次选择较小的集合。
            2. 如果这两个集合有交集，则访问结束
            - 时间复杂度：理论上和BFS是一样的，实际上可能会大量减少，毕竟会减少好多无用的探索。

```python
# 127 BFS  
# time 620ms beat 26%; 
# space 16.2MB beat 43%
from collections import deque, defaultdict
class Solution:
    def __init__(self):
        self.target = None
        self.word2children = defaultdict(list)

    def diff_one(self, node, other_node):
        return sum([node[i] != other_node[i] for i in range(len(node))]) == 1
    
    def buildGraph(self, beginWord, wordList):
        all_word = [beginWord] + wordList
        for i, node in enumerate(all_word):
            for other_node in wordList[i: ]:
                if self.diff_one(node, other_node):
                    self.word2children[node].append(other_node)
                    self.word2children[other_node].append(node)

    def findChildren(self, node):
        # find all unvisited children
        # implement two ways to findChildren: 
        # one way is to build a graph in advance and query directly 
        # the other way is to iterate all the possible candidates and find the valid ones.
        # choose according to the dict size with candidate variations size.
        
        children = []
        if self.word_len < 26:
            # build a graph in advance and query directly 
            all_children = self.word2children[node]
            for child in all_children:
                if not self.visited[child]:
                    children.append(child)
        else:
            # iterate all the possible candidates and find the valid ones
            for i in range(len(node)):
                for c in range(26):
                    if chr(ord('a') + c) == node[i]:
                        continue
                    var_node = node[:i] + chr(ord('a') + c) + node[i+1:]
                    if var_node in self.wordList and not self.visited[var_node]:
                        children.append(var_node)
        return children

    def BFS(self, begin_node):
        next_layer = [begin_node]
        path_len = 1
        while next_layer:
            cur_layer = next_layer
            next_layer = []
            for node in cur_layer:
                if self.diff_one(node, self.target):
                    return path_len + 1
                next_layer.extend(self.findChildren(node)) # find all unvisited children
            path_len += 1
        return 0

    def ladderLength(self, beginWord: str, endWord: str, wordList: List[str]) -> int:
        if endWord not in wordList:
            return 0
        self.target = endWord
        self.wordList = set(wordList)
        self.visited = {word: False for word in self.wordList}
        self.visited[beginWord] = True
        self.word_len = len(wordList)
        if self.word_len < 26:
            self.buildGraph(beginWord, wordList)
        return self.BFS(beginWord)
```

```python
# 127 双向BFS
# time 96ms beat 91.76%; 
# space 16.3MB beat 38.43%

from collections import deque, defaultdict
class Solution:
    def __init__(self):
        self.target = None
        self.word2children = defaultdict(list)

    def diff_one(self, node, other_node):
        return sum([node[i] != other_node[i] for i in range(len(node))]) == 1
    
    def buildGraph(self, beginWord, wordList):
        all_word = [beginWord] + wordList
        for i, node in enumerate(all_word):
            for other_node in wordList[i: ]:
                if self.diff_one(node, other_node):
                    self.word2children[node].append(other_node)
                    self.word2children[other_node].append(node)

    def findChildren(self, node, layer_id):
        children = []
        if self.word_len < 26:
            all_children = self.word2children[node]
            for child in all_children:
                if not self.visited[layer_id][child]:
                    children.append(child)
                    self.visited[layer_id][child] = True
        else:
            for i in range(len(node)):
                for c in range(26):
                    if chr(ord('a') + c) == node[i]:
                        continue
                    var_node = node[:i] + chr(ord('a') + c) + node[i+1:]
                    if var_node in self.wordList and not self.visited[layer_id][var_node]:
                        children.append(var_node)
                        self.visited[layer_id][var_node] = True
        return children

    def BiBFS(self, begin_node, end_node):
        self.next_layers = {} # 双向的记录
        self.next_layers[0] = [begin_node]
        self.next_layers[1] = [end_node]
        cur_layer_id = 0 # 每次选择较小的方向进行展开
        path_len = 2
        while self.next_layers[cur_layer_id]:
            cur_layer = self.next_layers[cur_layer_id]
            self.next_layers[cur_layer_id] = []
            for node in cur_layer:
                children = self.findChildren(node, cur_layer_id) # 先找到所有本方向没有访问过的点
                for child in children:
                    if self.visited[1-cur_layer_id][child]: # 再判断其是否在另一个方向被访问过
                        return path_len
                self.next_layers[cur_layer_id].extend(children)
            path_len += 1
            if len(self.next_layers[cur_layer_id]) > len(self.next_layers[1-cur_layer_id]):
                cur_layer_id = 1 - cur_layer_id # 每次选择候选较少的方向展开
        return 0

    def ladderLength(self, beginWord: str, endWord: str, wordList: List[str]) -> int:
        if endWord not in wordList:
            return 0
        self.target = endWord
        self.wordList = set(wordList)
        self.visited = defaultdict(dict)
        for word in self.wordList:
            self.visited[0][word] = False # 双向的记录
            self.visited[1][word] = False
        self.visited[0][beginWord] = True
        self.visited[1][endWord] = True
        self.word_len = len(wordList)
        if self.word_len < 26:
            self.buildGraph(beginWord, wordList)
        return self.BiBFS(beginWord, endWord)
```



## 最短路

https://dowalle.gitbook.io/algo/algorithm/4-tu-lun/7-zui-duan-lu-jing

- [(medium)1631Path With Minimum Effort](https://leetcode.cn/problems/path-with-minimum-effort/)。这道题是种求最短路的变式。一般而言，求最短路可以尝试BFS。知道目标的时候，也可以尝试双向BFS
    - 思路一：BFS
    - 思路二：Dijkstra.思考以下问题：
        - [X] 满足什么条件时，可以用dijkstra算法？什么时候不可以？127这样的题目可以用吗？当候选的多个点之间，有“优劣”之分的时候，选择dijkstra算法比较有优势。127的候选节点没有优劣之分，所以d算法并不会带来优势，故不选用。
        - 复杂度分析：Dijkstra算法中，每条边至多被选择一遍（复杂度是O(m)的，假设有m条边），且每次选择当前最优的一条边，复杂度是O(logm)的。总的而言时间复杂度是O(mlogm)的，空间复杂度是O(m)的，用于存储待选择的边。在本题中，一个m * n的矩阵，有(m-1)* n + (n-1)*m = O(mn)条边。故时间复杂度是O(mn log(mn))的，空间复杂度是O(mn)的。
    - [X] 思路三：并查集。并查集的做法，就是看头和尾什么时候能连接到一起，每次选取最省力的那条边，来连接节点们。当头和尾连接到一起的时候，就是结果。这里一个前提是，我们知道头和尾一定是可以连到一起的，关键就在于找最短路。并查集需要枚举出所有的边(O(mn))，且在每次取出最省力的那一条(O(logmn)，如果用堆的话)。总体而言时间复杂度是O(mnlog(mn))的，空间复杂度确定是O(mn)的。

```python
# 1631 BFS
# time 3136ms
# space 16.2MB
class Solution:
    def findNext(self, i, j, heights):
        nexts = []
        for next_i, next_j in [[i-1, j], [i+1, j], [i, j - 1], [i, j + 1]]:
            if next_i < 0 or next_i >= self.m or next_j < 0 or next_j >= self.n:
                continue
            cur_effort = max(self.efforts[i][j], abs(heights[next_i][next_j] - heights[i][j])) # 判断是否要继续访问
            if self.efforts[next_i][next_j] == -1 or self.efforts[next_i][next_j] > cur_effort:
                self.efforts[next_i][next_j] = cur_effort
                nexts.append((next_i, next_j))
        return nexts

    def BFS(self, heights):
        next_layer = [(0, 0)]
        while next_layer:
            cur_layer = next_layer
            next_layer = []
            for i, j in cur_layer:
                next_layer.extend(self.findNext(i, j, heights))
        return

    def minimumEffortPath(self, heights: List[List[int]]) -> int:
        self.m = len(heights)
        self.n = len(heights[0])
        self.efforts = [[-1 for _ in range(self.n)] for _ in range(self.m)]
        self.efforts[0][0] = 0
        self.BFS(heights)
        return self.efforts[-1][-1]
```

```python
# 1631 并查集
# time 776ms, beat 74.51%
# space 19MB, beat 21.89%

import heapq
class UnionSet:
    def __init__(self, n):
        self.root = {}
        for i in range(n):
            self.root[i] = i
    def findRoot(self, i):
        while i != self.root[i]:
            i = self.root[i]
        return i

    def add(self, i, j):
        i = self.findRoot(i)
        j = self.findRoot(j)
        if i > j:
            i, j = j, i
        self.root[j] = self.root[i]
    
    def connected(self, i, j):
        i = self.findRoot(i)
        j = self.findRoot(j)
        return j == i

class Solution:
    def minimumEffortPath(self, heights: List[List[int]]) -> int:
        # build a union set for each node
        m = len(heights)
        n = len(heights[0])
        if m * n == 1:
            return 0
        union_set = UnionSet(m*n)
        # add all the edges into a heap
        candidates = []
        for i in range(m):
            for j in range(n):
                for next_i, next_j in [[i+1, j], [i, j+1]]:
                    if next_i >= m or next_j >= n:
                        continue
                    effort = abs(heights[next_i][next_j] - heights[i][j])
                    heapq.heappush(candidates, (effort, next_i * n + next_j, i * n + j))
        # each time pop an edge from the heap and add it to the union set.
        while candidates:
            best_cand = heapq.heappop(candidates)
            union_set.add(best_cand[1], best_cand[2])   
            # if the start and end are in the same set, return the effort
            if union_set.connected(0, m * n - 1):
                return best_cand[0]
        return -1
```

```python
# 1631 二分查找
# time 2832ms, beat 21.57%
# space 15.8MB, beat 97.39%

import heapq
class Solution:
    def isReachable(self, mid, heights):
        visited = [[False for _ in range(self.n)] for _ in range(self.m)]
        next_layer = [[0, 0]]
        visited[0][0] = True
        while next_layer:
            cur_layer = next_layer
            next_layer = []
            for i, j in cur_layer:
                if i == self.m - 1 and j == self.n - 1:
                    return True
                for next_i, next_j in [[i - 1, j], [i + 1, j], [i, j - 1], [i, j + 1]]:
                    if next_i < 0 or next_i >= self.m or next_j < 0 or next_j >= self.n or visited[next_i][next_j] or abs(heights[i][j] - heights[next_i][next_j]) > mid:
                        continue
                    visited[next_i][next_j] = True
                    next_layer.append([next_i, next_j])
        return False

    def minimumEffortPath(self, heights: List[List[int]]) -> int:
        # binary search for the possible height
        self.m = len(heights)
        self.n = len(heights[0])
        # print("m: ", self.m, " n: ", self.n)
        right = max([max(l) for l in heights])
        left = 0
        # print(left, right)
        while left < right:
            mid = (left + right) // 2
            if self.isReachable(mid, heights):
                right = mid
            else:
                left = mid + 1
        return right
```

```python
# 210 的四种实现（前/后， BFS/DFS）
# 前 + B
# time 44ms, beat 70.84%
# space 16MB, beat 54.79%

# 前 + D
# time 44ms, beat 70.84%
# space 17.8MB, beat 30.30%

# 后 + B
# time 44ms, beat 70.84%
# space 16.2MB, beat 45.46%

# 后 + D
# time 48ms, beat 47.53%
# space 17.9MB, beat 22.91%

from collections import defaultdict
class Solution:
    def ForwardBFS(self, start_nodes):
        next_layer = start_nodes
        res = []
        while next_layer:
            cur_layer = next_layer
            next_layer = []
            for course in cur_layer:
                res.append(course)
                later_courses = self.course2later.get(course, [])
                for later_c in later_courses:
                    self.in_degree[later_c] -= 1
                    if self.in_degree[later_c] == 0:
                        next_layer.append(later_c)
        return res

    def ForwardDFS(self, start):
        self.res.append(start)
        for later_c in self.course2later.get(start, []):
            self.in_degree[later_c] -= 1
            if self.in_degree[later_c] == 0:
                self.ForwardDFS(later_c)

    def BackwordBFS(self, ends):
        res = []
        next_layer = ends
        while next_layer:
            cur_layer = next_layer
            next_layer = []
            for course in cur_layer:
                res.append(course)
                for pre in self.course2pre.get(course, []):
                    self.out_degree[pre] -= 1
                    if self.out_degree[pre] == 0:
                        next_layer.append(pre)
        return res[::-1]

    def BackwordDFS(self, node):
        self.res.append(node)
        for pre in self.course2pre.get(node, []):
            self.out_degree[pre] -= 1
            if self.out_degree[pre] == 0:
                self.BackwordDFS(pre)


    def findOrder(self, numCourses: int, prerequisites: List[List[int]]) -> List[int]:
        direction = "F" # "B"
        search = 'B' # or "D"
        if direction == "F":
            self.in_degree = [0 for i in range(numCourses)]
            self.course2later = defaultdict(list)
            for pair in prerequisites:
                self.in_degree[pair[0]] += 1
                self.course2later[pair[1]].append(pair[0])
            starts = []
            for i in range(numCourses):
                if self.in_degree[i] == 0:
                    starts.append(i)
            if not starts:
                return []
            if search == "B":
                res = self.ForwardBFS(starts)
                return res if len(res) == numCourses else []
            else:
                self.res = []
                for start in starts:
                    self.ForwardDFS(start)
                return self.res if len(self.res) == numCourses else []
        else:
            self.out_degree = [0 for _ in range(numCourses)]
            self.course2pre = defaultdict(list)
            for pair in prerequisites:
                self.out_degree[pair[1]] += 1
                self.course2pre[pair[0]].append(pair[1])
            ends = []
            for i in range(numCourses):
                if self.out_degree[i] == 0:
                    ends.append(i)
            if not ends:
                return []
            if search == "B":
                res = self.BackwordBFS(ends)
                return res if len(res) == numCourses else []
            else:
                self.res = []
                for end in ends:
                    self.BackwordDFS(end)
            return self.res[::-1] if len(self.res) == numCourses else []
```
