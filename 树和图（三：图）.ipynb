```python
# 785 å¤šæ¬¡BFS
class Solution:
    def __init__(self):
        self.color = []

    def getUncolored(self):
        res = []
        for i, c in enumerate(self.color):
            if c == 0:
                res.append(i)
        return res
    
    def coloring(self, start_node, color):
        next2color = set([start_node])
        while next2color:
            cur2color = next2color
            next2color = set()
            for node in cur2color:
                self.color[node] = color
                for neigh in self.graph[node]:
                    if self.color[neigh] == 0:
                        next2color.add(neigh)
                    elif self.color[neigh] == color:
                        return False
            color = 3 - color
        return True

    def isBipartite(self, graph: List[List[int]]) -> bool:
        self.color = [0] * len(graph)
        self.graph = graph
        while True:
            uncolored_nodes = self.getUncolored() # è¿™ä¸ªæ˜¯å¼€å¯å¤šæ¬¡æŸ“è‰²è¿‡ç¨‹
            if not uncolored_nodes:
                return True
            start_node = uncolored_nodes[0]
            if not self.coloring(start_node, 1): # æœ¬æ¬¡æŸ“è‰²æ˜¯å¦æˆåŠŸ
                return False
        return False
```

```python
# 1631 BFS
# time 3136ms
# space 16.2MB
class Solution:
    def findNext(self, i, j, heights):
        nexts = []
        for next_i, next_j in [[i-1, j], [i+1, j], [i, j - 1], [i, j + 1]]:
            if next_i < 0 or next_i >= self.m or next_j < 0 or next_j >= self.n:
                continue
            cur_effort = max(self.efforts[i][j], abs(heights[next_i][next_j] - heights[i][j])) # åˆ¤æ–­æ˜¯å¦è¦ç»§ç»­è®¿é—®
            if self.efforts[next_i][next_j] == -1 or self.efforts[next_i][next_j] > cur_effort:
                self.efforts[next_i][next_j] = cur_effort
                nexts.append((next_i, next_j))
        return nexts

    def BFS(self, heights):
        next_layer = [(0, 0)]
        while next_layer:
            cur_layer = next_layer
            next_layer = []
            for i, j in cur_layer:
                next_layer.extend(self.findNext(i, j, heights))
        return

    def minimumEffortPath(self, heights: List[List[int]]) -> int:
        self.m = len(heights)
        self.n = len(heights[0])
        self.efforts = [[-1 for _ in range(self.n)] for _ in range(self.m)]
        self.efforts[0][0] = 0
        self.BFS(heights)
        return self.efforts[-1][-1]
```

```python
# 210 çš„å››ç§å®ç°ï¼ˆå‰/åï¼Œ BFS/DFSï¼‰
# å‰ + B
# time 44ms, beat 70.84%
# space 16MB, beat 54.79%

# å‰ + D
# time 44ms, beat 70.84%
# space 17.8MB, beat 30.30%

# å + B
# time 44ms, beat 70.84%
# space 16.2MB, beat 45.46%

# å + D
# time 48ms, beat 47.53%
# space 17.9MB, beat 22.91%

from collections import defaultdict
class Solution:
    def ForwardBFS(self, start_nodes):
        next_layer = start_nodes
        res = []
        while next_layer:
            cur_layer = next_layer
            next_layer = []
            for course in cur_layer:
                res.append(course)
                later_courses = self.course2later.get(course, [])
                for later_c in later_courses:
                    self.in_degree[later_c] -= 1
                    if self.in_degree[later_c] == 0:
                        next_layer.append(later_c)
        return res

    def ForwardDFS(self, start):
        self.res.append(start)
        for later_c in self.course2later.get(start, []):
            self.in_degree[later_c] -= 1
            if self.in_degree[later_c] == 0:
                self.ForwardDFS(later_c)

    def BackwordBFS(self, ends):
        res = []
        next_layer = ends
        while next_layer:
            cur_layer = next_layer
            next_layer = []
            for course in cur_layer:
                res.append(course)
                for pre in self.course2pre.get(course, []):
                    self.out_degree[pre] -= 1
                    if self.out_degree[pre] == 0:
                        next_layer.append(pre)
        return res[::-1]

    def BackwordDFS(self, node):
        self.res.append(node)
        for pre in self.course2pre.get(node, []):
            self.out_degree[pre] -= 1
            if self.out_degree[pre] == 0:
                self.BackwordDFS(pre)


    def findOrder(self, numCourses: int, prerequisites: List[List[int]]) -> List[int]:
        direction = "F" # "B"
        search = 'B' # or "D"
        if direction == "F":
            self.in_degree = [0 for i in range(numCourses)]
            self.course2later = defaultdict(list)
            for pair in prerequisites:
                self.in_degree[pair[0]] += 1
                self.course2later[pair[1]].append(pair[0])
            starts = []
            for i in range(numCourses):
                if self.in_degree[i] == 0:
                    starts.append(i)
            if not starts:
                return []
            if search == "B":
                res = self.ForwardBFS(starts)
                return res if len(res) == numCourses else []
            else:
                self.res = []
                for start in starts:
                    self.ForwardDFS(start)
                return self.res if len(self.res) == numCourses else []
        else:
            self.out_degree = [0 for _ in range(numCourses)]
            self.course2pre = defaultdict(list)
            for pair in prerequisites:
                self.out_degree[pair[1]] += 1
                self.course2pre[pair[0]].append(pair[1])
            ends = []
            for i in range(numCourses):
                if self.out_degree[i] == 0:
                    ends.append(i)
            if not ends:
                return []
            if search == "B":
                res = self.BackwordBFS(ends)
                return res if len(res) == numCourses else []
            else:
                self.res = []
                for end in ends:
                    self.BackwordDFS(end)
            return self.res[::-1] if len(self.res) == numCourses else []
```
