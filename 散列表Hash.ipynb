散列表（哈希表）本质是一个key-value存储的“数组”。对于查找很有帮助！！！一种空间换时间的数据结构。
- 最直接的形式：对于要存储的对象value，使用一个哈希函数，将其key映射到一个内存单元。存储时，直接在该内存单元存储value。
- 变式（hashset）：value本身作为key，使用一个哈希函数，将其key映射到一个内存单元。存储时，直接在该内存单元存储1。删除时，在该内存单元存储0（或其他用于区分的值）

这里有三个对象：
- key：是唯一标识这个数据对象的，可以是这个数据对象本身，也可以是其中的部分属性。
    - [X] key是如何选取的。简单的数据类型，常为本身；复杂的，可以人为构造id等。在python中不可变对象是可哈希的，不管这个对象有多复杂，不可变意味着不同对象的地址不一样，这时就可以用它们的地址来求哈希值啦。
- value：要存储的完整的数据对象
- 内存单元：内存单元存储的是value。也不是绝对的，比如python的dict实现，字典的每一个键都占用一个单元，一个单元分为两部分，分别是对键的引用和对值的引用，使用hash函数获取键的散列值，散列值对数组长度取余，取得值就是存放位置的索引。这种存放key的好处是，便于rehash（扩张、缩减）

## 哈希表实现方法
假设要存储的数据元素个数为n，设置一个长度为m（m≥n）的连续存储单元，分别以每个数据元素的关键字 Ki(0<= i <=n-1) 为自变量，通过哈希函数 hash(Ki) 把 Ki 映射为内存单元的某个地址 hash(Ki)，并将该数据元素存储在该内存单元中。这实际上是一种数组存储的实现。

- [X] 所以，实际上哈希表需要的内存空间是大于实际要存储的数据个数的？要大多少呢？
    是的，一般哈希表在初始化的时候都有一个默认的大小。当这个哈希表被填充达到一定比例的时候，会扩大size，执行rehash。大多少不一定，由具体实现（比如规定的填充比例）来决定。
- [X] 这是不是说，我需要根据对象的大小来确定内存单元的大小？如果各数据对象value的size大小不一样怎么办？
    不是的，哈希表的内存单元实际存储的不是value，而是一个指向value的指针，这个指针的size是统一的。



底层实现的各种形式：
- 最基本：数组：一块连续的内存单元，将key得到的哈希值按照哈希表的size取模即为存储地址。缺点是：无法解决哈希碰撞，为了减少哈希碰撞，需要非常多的额外空间
- 进阶：数组 + 数组/链表：为了解决哈希碰撞的一种扩展变化：当两个key映射到同一个内存单元时，可以通过额外的存储来同时存放。
    - 数组：可以看成是多层哈希。首先得到的哈希值，定位到某个子数组；在子数组里继续哈希，得到具体位置。以此类推，可根据需要多层嵌套。缺点是：大量的额外空间
    - 链表：好处是插入操作的复杂度为O(1)，缺点是查找为O(n)。
- 高阶：数组 + 链表/红黑树：当某个桶（最外层的内层单元）内存放了比较多的数值时，使用链表的查找复杂度太高，这时可以使用一个高度平衡的二叉树来替代。好处是将查找的复杂度降为O(logN)，但是把插入的复杂度也提升到了O(logN)


各种应用
- HashMap：使用key存储value。作为python中dict的底层实现。对外接口
    - add(key, val)新增一个key-val对
    - delete(key)删除该key及其对应的value
    - find(key)判断key是否存在
- HashSet：存储元素item，并快速判断一个item是否在当前集合中。作为python中set的底层实现。对外接口
    - add(item)新增一个item
    - delete(item)删除该item
    - find(item)判断item是否存在

```python
# Version 1.0 HashMap

class HashMap(object):
    def __init__(self, init_size=100): # init_size:初始化的数组大小
        self._size = init_size
        self._table = [None for _ in range(self._size)]
    
    def hash_func(self, key): # 规定使用的哈希函数，这里使用最基本的：直接取模
        assert key is int
        return key % self._size
        
    def add(self, key, val):
        loc = self.hash_func(key)
        self._table[loc] = val
        
    def delete(self, key):
        loc = self.hash_func(key)
        self._table[loc] = None
    
    def find(self, key):
        loc = self.hash_func(key)
        return self._table[loc] # if exists, return the value; otherwise, return None
        
        
# Version 1.0 HashSet

class HashSet(object):
    def __init__(self, init_size=100): # init_size:初始化的数组大小
        self._size = init_size
        self._table = [0 for _ in range(self._size)]
    
    def hash_func(self, key): # 规定使用的哈希函数，这里使用最基本的：直接取模
        assert key is int
        return key % self._size
        
    def add(self, value):
        loc = self.hash_func(value)
        self._table[loc] = 1 # 和map不一样的地方在于，存储的不是value了，而是0或1用来标志是否存在。
        
    def delete(self, value):
        loc = self.hash_func(value)
        self._table[loc] = 0
    
    def find(self, value):
        loc = self.hash_func(value)
        return self._table[loc] == 1 # if exists, return True; otherwise, return False
        
        
```

```python
# Version 2.0 HashMap, 应对碰撞

class HashMap(object):
    def __init__(self, init_size=100, sub_size=100):
        self._size = init_size
        self._sub_size = sub_size # 使用一个数组/链表，在碰撞时存储多个值。更高阶的3.0等，这里可以使用一个二叉树/红黑树。
        self._table = [[None for _ in range(self._sub_size)] for _ in range(self._size)]
    
    def hash_func(self, key, size):
        assert key is int
        return key % _size
        
    def add(self, key, val):
        loc = self.hash_func(key, self._size)
        sub_loc = self.hash_func(key, self._sub_size)
        self._table[loc][sub_loc] = val
        
    def delete(self, key):
        loc = self.hash_func(key)
        sub_loc = self.hash_func(key, self._sub_size)
        self._table[loc][sub_loc] = None
    
    def find(self, key):
        loc = self.hash_func(key)
        sub_loc = self.hash_func(key, self._sub_size)
        return self._table[loc][sub_loc]
        
        
# Version 2.0 HashSet

class HashSet(object):
    def __init__(self, init_size=100, sub_size=100): 
        self._size = init_size
        self._sub_size = sub_size # 使用一个数组/链表，在碰撞时存储多个值
        self._table = [[0 for _ in range(self._sub_size)] for _ in range(self._size)]
    
    def hash_func(self, key, size):
        assert key is int
        return key % size
        
    def add(self, value):
        loc = self.hash_func(value)
        sub_loc = self.hash_func(key, self._sub_size)
        self._table[loc][sub_loc] = 1
        
    def delete(self, value):
        loc = self.hash_func(value)
        sub_loc = self.hash_func(key, self._sub_size)
        self._table[loc][sub_loc] = 0
    
    def find(self, value):
        loc = self.hash_func(value)
        sub_loc = self.hash_func(key, self._sub_size)
        return self._table[loc][sub_loc] == 1 
        
```


## 时间空间复杂度


整体而言，哈希表的空间复杂度是O(n)，即随着存储元素数量线性增长。

哈希表的时间复杂度与设计有很强的关系。

数组+链表/数组的设计：
我们中的大多数人可能已经在每个桶中使用数组来将值存储在同一个桶中，
- 理想情况下，桶的大小足够小时，可以看作是一个常数。插入和搜索的时间复杂度都是 O(1)。
- 最坏的情况下，桶大小的最大值将为N。插入时时间复杂度为 O(1)，搜索时为 O(N)。

数组+链表+二叉树/红黑树的设计：

如果在同一个桶中有太多的值，这些值将被保留在一个高度平衡的二叉树搜索树中。
- 插入和搜索的平均时间复杂度仍为 O(1)。
- 最坏情况下插入和搜索的时间复杂度是 O(logN)

## 哈希冲突/碰撞及解决策略

哈希冲突/碰撞：一言以蔽之就是不同的key被hash到了同一个位置。

解决策略：
- 链接法：也就是上述实现里的2.0版本，说白了就是如果冲突了，就都存下来。是对内存的考验
- 开放寻址法：通过在数组以某种方式寻找数组中**空余的结点放置**。基本思想是：当关键字key的哈希地址p=H（key）出现冲突时，以p为基础，产生另一个哈希地址p1，如果p1仍然冲突，再以p为基础，产生另一个哈希地址p2，…，直到找出一个不冲突的哈希地址pi
    - [X] 如何进行查找呢？能实现查找的前提条件是：存放的内存单元里，有key的部分（比如是python的字典实现：存放了key的索引）。这样沿着索引序列查找时，如果一个内存单元已经被占用，可以通过比较里面存储的key是不是查询的key，来决定是否查找到，以及要不要继续。直到遇到了空白单元，则查找失败。
    - [X] 如和进行删除呢？[reference](https://www.cnblogs.com/east7/p/12594894.html)不能真正的删除，而是做一个记号，否则会引起查找错误。只有当该内存被下一次填充的时候，才算是真正的删除了。

```python
# 030 
# 参考题解
class Solution:
    def findSubstring(self, s: str, words: List[str]) -> List[int]:
        from collections import Counter
        if not s or not words:return []
        one_word = len(words[0])
        word_num = len(words)
        n = len(s)
        if n < one_word:return []
        words = Counter(words)
        res = []
        for i in range(0, one_word): # 最外层循环：次数为单词的长度
            cur_cnt = 0
            left = i
            right = i
            cur_Counter = Counter()
            while right + one_word <= n:
                w = s[right:right + one_word]
                right += one_word
                if w not in words: # 当前单词不在给定列表里时，清空重来
                    left = right
                    cur_Counter.clear()
                    cur_cnt = 0
                else:
                    cur_Counter[w] += 1
                    cur_cnt += 1
                    while cur_Counter[w] > words[w]: # 如果某个单词的出现次数超过了给定的次数，则删除之前出现。这里有个冗余操作。
                        left_w = s[left:left+one_word]
                        left += one_word
                        cur_Counter[left_w] -= 1
                        cur_cnt -= 1
                    if cur_cnt == word_num :
                        res.append(left)
        return res
```

```python
# 记录下标的题解
from collections import Counter
class Solution:
    def findSubstring(self, s: str, words: List[str]) -> List[int]:
        words_num = Counter(words)
        start_idx = []
        unit_size = len(words[0])
        for start_i in range(unit_size):
            start_id = start_i # 用于记录当前有效的起始位置
            words_loc = {} # 记录各个单词出现的位置
            for i in range(start_i, len(s), unit_size):
                cur_word = s[i: (i+unit_size)] 
                if cur_word not in words_num:
                    start_id = i + unit_size
                    words_loc = {}
                    continue
                if cur_word in words_loc:
                    new_loc = []
                    for valid_id, saved_loc in enumerate(words_loc[cur_word]): # 只保留在当前起始位置之后的记录
                        if saved_loc >= start_id:
                            new_loc = words_loc[cur_word][valid_id: ]
                            break
                    new_loc.append(i)
                    words_loc[cur_word] = new_loc
                    if len(new_loc) > words_num[cur_word]:
                        start_id = words_loc[cur_word][0] + unit_size
                        words_loc[cur_word] = words_loc[cur_word][1:]
                else:
                    words_loc[cur_word] = [i]

                # judge if the current words_loc satisfies the condition
                if i - start_id == unit_size * (len(words)-1):
                    start_idx.append(start_id)
                    start_id += unit_size
        return start_idx
```

```python
# 076 我的题解
"""
思路是：
- 用滑动窗口来解，控制时间复杂度为O(n)
- 滑动窗口维护的条件是：窗口中包含t的字符数目<t的长度。一旦相等，先移动窗口至其“最小”，再判断并保存结果
- 这里复杂的一点是：一个num记录的是不超过t中的字符出现次数，一个counter记录的是每个字符实际出现的次数，用于缩短窗口。
- 移动窗口至其“最小”的条件：再多移动一个，就破坏了成立条件。

"""


from collections import Counter
class Solution:
    def minWindow(self, s: str, t: str) -> str:
        if len(t) > len(s):
            return ""
        min_str = ""
        min_len = 0
        start_id = 0
        existed_num = 0
        ttl_num = len(t)
        t_cnt = Counter([c for c in t])
        cur_cnt = Counter()
        for end_id, end_char in enumerate(s):  # 用滑动窗口来解
            if end_char not in t:
                continue
            if cur_cnt[end_char] < t_cnt[end_char]:
                existed_num += 1
            cur_cnt[end_char] += 1
            if existed_num == ttl_num:
                while existed_num == ttl_num: # 移动窗口，缩短左边，减小字符串长度
                    if s[start_id] in t:
                        cur_cnt[s[start_id]] -= 1
                        if cur_cnt[s[start_id]] < t_cnt[s[start_id]]:
                            existed_num -= 1
                    start_id += 1
                if min_len == 0 or (end_id - start_id + 2) < min_len: # 实际的其实位置是start_id - 1，start_id其实是第一次破坏了窗口的右边一位的下标
                    min_len = end_id - start_id + 2
                    min_str = s[(start_id - 1): (end_id + 1)]
        return min_str
```

```python
# 128
class Solution:
    def longestConsecutive(self, nums: List[int]) -> int:
        max_len = 0
        max_lengths = dict()
        for num in nums:
            if num in max_lengths:
                continue
            left = max_lengths.get(num - 1, 0)
            right = max_lengths.get(num + 1, 0)
            cur_len = left + 1 + right
            max_len = max(max_len, cur_len)
            # print("Before update: ", max_lengths)
            max_lengths[num - left] = cur_len
            max_lengths[num + right] = cur_len
            max_lengths[num] = cur_len
            # print("After update: ",max_lengths)
        return max_len
```

```python
# 720
class Solution:
    def order_word(self, words):
        max_len = max([len(w) for w in words])
        len2words = [[] for _ in range(max_len + 1)]
        len2words[0].append('')
        for word in words:
            len2words[len(word)].append(word)
        return max_len, len2words
    def check_words(self, max_len, len2words):
        valid_words = [['']]
        for l in range(1, max_len+1):
            cur_words = []
            for word in len2words[l]:
                if word[:-1] in valid_words[-1]:
                    cur_words.append(word)
            if len(cur_words) == 0:
                return valid_words
            valid_words.append(sorted(cur_words))
        return valid_words
        


    def longestWord(self, words: List[str]) -> str:
        max_len, len2words = self.order_word(words) 
        valid_words = self.check_words(max_len, len2words)
        return valid_words[-1][0]
```

```python
# 726

from collections import Counter
class Solution:
    def split_formula(self, formula): # return a list of items like (num, str), num: the frequence of the str
        if "(" not in formula:
            return [[1, formula]]
        splited = []
        left_cnt = 0
        right_cnt = 0
        within = True
        num_str = ""
        start_id = 0
        for end_id, cur_char in enumerate(formula):
            if cur_char == "(":
                if left_cnt == 0 and start_id != end_id:
                    num = 1 if num_str == "" else int(num_str)
                    if formula[start_id] == "(":
                        splited.append([num, formula[start_id + 1: end_id - len(num_str) - 1]])
                    else:
                        splited.append([num, formula[start_id: end_id - len(num_str)]])
                    start_id = end_id
                    num_str = ""
                    within = True
                left_cnt += 1
            elif cur_char == ")":
                left_cnt -= 1
                if left_cnt == 0:
                    within = False
            elif cur_char.isdigit() and not within:
                num_str += cur_char
            else:
                if cur_char.isupper() and (num_str != "" or formula[end_id - 1] == ")"):
                    num = 1 if num_str == "" else int(num_str)
                    if formula[start_id] == "(":
                        splited.append([num, formula[start_id + 1: end_id - len(num_str) - 1]])
                    else:
                        splited.append([num, formula[start_id: end_id - len(num_str)]])
                    start_id = end_id
                    num_str = ""
                    within = True

                continue
        num = 1 if num_str == "" else int(num_str)
        if formula[start_id] == "(":
            splited.append([num, formula[start_id + 1: end_id - len(num_str)]])
        else:
            splited.append([num, formula[start_id: end_id - len(num_str) + 1]])
        return splited

    def isBasicFormula(self, formula):
        return "(" not in formula

    def splitbyBigAlpha(self, formula):
        alpha_list = []
        start_id = 0
        for i in range(1, len(formula)):
            if formula[i].isupper():
                alpha_list.append(formula[start_id: i])
                start_id = i
        alpha_list.append(formula[start_id: ])
        return alpha_list
    
    def extract_number(self, alpha):
        pure_alpha = alpha
        num = 1
        for i in range(len(alpha)):
            if alpha[i].isdigit():
                pure_alpha = alpha[:i]
                num = int(alpha[i:])
                break
        return num, pure_alpha

    def calcBasic(self, formula):
        alpha_list = self.splitbyBigAlpha(formula)
        counter = Counter()
        # print(alpha_list)
        for alpha in alpha_list:
            num, pure_alpha = self.extract_number(alpha)
            counter[pure_alpha] += num
        return counter

    def dictCount(self, formula):
        count = Counter()
        formula_list = self.split_formula(formula)
        for form in formula_list:
            if self.isBasicFormula(form[1]):
                cur_count = self.calcBasic(form[1])
            else:
                cur_count = self.dictCount(form[1])
            for key, val in cur_count.items():
                cur_count[key] = val * form[0]
            count.update(cur_count)
        return count

    def Counter2Str(self, counter):
        res = ""
        keys = sorted(counter.keys())
        for key in keys:
            val = counter[key]
            if val == 1:
                res += key
            else:
                res += (key + str(val))
        return res

    def countOfAtoms(self, formula: str) -> str:
        counter = self.dictCount(formula)
        return self.Counter2Str(counter)
```

```python
# 930
class Solution:
    def numSubarraysWithSum(self, nums: List[int], goal: int) -> int:
        cnt = 0
        l1 = 0
        l2 = 0
        s1 = 0
        s2 = 0
        for r in range(len(nums)):
            s1 += nums[r]
            s2 += nums[r]
            while s1 > goal and l1 <= r:
                s1 -= nums[l1]
                l1 += 1
            while s2 >= goal and l2 <= r:
                s2 -= nums[l2]
                l2 += 1
            if s1 == goal: # 这个判断条件可以省略。
                cnt += (l2 - l1)
        return cnt
```

```python
# 992
class Solution:
    def subarraysWithKDistinct(self, nums: List[int], k: int) -> int:
        res = 0
        l1 = 0
        l2 = 0 # the ideal window's left side is within [l1, l2), [l1, r]
        l1_diff = 0
        l2_diff = 0
        l1_count = dict()
        l2_count = dict()
        for r, num in enumerate(nums):
            l1_count[num] = l1_count.get(num, 0) + 1
            l2_count[num] = l2_count.get(num, 0) + 1
            if l1_count[num] == 1:
                l1_diff += 1
            if l2_count[num] == 1:
                l2_diff += 1
            while l2_diff == k:
                l2_count[nums[l2]] -= 1
                if l2_count[nums[l2]] == 0:
                    l2_diff -= 1
                l2 += 1
            while l1_diff == k + 1:
                l1_count[nums[l1]] -= 1
                if l1_count[nums[l1]] == 0:
                    l1_diff -= 1
                l1 += 1
            if l1_diff == k: # without the last item, the diff num is k.
                res += l2 - l1
        return res
```

# 实际应用

- 哈希表是python中dict结构，set结构的底层实现
- 某些id的生成方式使用的是murmurhash函数（利用均匀性，可逆）
- 某些kv数据库的实现：redis的存储方式？http://redisbook.com/preview/dict/rehashing.html
- 在比特币中的应用：使用的哈希函数是SHA-256（Secure Hash Algorithm，安全散列算法）。（不可逆）
- 基于hash生成的特征值：LSH（local sensitive hash）局部敏感哈希：设计一种哈希函数对这两点进行哈希值计算，使得他们哈希值有很大的概率是一样的。同时若两点之间的距离较远，他们哈希值相同的概率会很小。（不均匀，相似相近）
    [思想](https://www.jianshu.com/p/d4368c8f40cb)：如果我们能先粗略地将用户分桶，将可能相似的用户以较大概率分到同一个桶内，这样每一个用户的“备选相似用户集”就会相对较小，降低寻找其相似用户的计算复杂度，LSH就是这样一个近似算法。核心思想是**哈希分桶**
- 布隆过滤器：多个独立的哈希函数，可用于快速检索。有误判，遵循“宁可错杀，不可误放”的原则。
